## <span style="color:blue">ESERCIZIO 7.1</span>

Prima di procedere con lo studio, è necessario valutare in quanti step il sistema raggiunga l'equilibrio. Si è deciso, quindi, di inserire in "input.dat" una variabile che, se pari ad uno, stampa su dei file i valori istantanei di pressione ed energia (per particella). Partendo dal reticolo perfetto, con i parametri dello stato liquido si sono ottenuti i seguenti risultati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

energia=np.loadtxt("eq_istantanei_energia.dat", usecols=(0), delimiter=' ', unpack='true')
pres=np.loadtxt("eq_istantanei_pressione.dat", usecols=(0), delimiter=' ', unpack='true')

x=np.arange(len(energia))
plt.plot(x,energia)
plt.title("Valori istantanei energia potenziale per particella")
plt.xlabel("step")
plt.ylabel("$U/N^*_{ist}$")
plt.grid(True)
plt.show()

plt.plot(x,pres)
plt.title("Valori istantanei pressione")
plt.xlabel("step")
plt.ylabel("$P^*_{ist}$")
plt.grid(True)
plt.show()
```

Si può notare che, in circa $200$ passi Monte Carlo, si raggiunga l'equilibrio. Utilizzando 1000 step iniziali prima di studiare il sistema, si può quindi essere ragionevolmente sicuri di campionare correttamente le distribuzioni di probabilità per ciascuno dei tre stati di aggregazione.

In analogia con il programma di dinamica molecolare, è stata anche inserita la possibilità di ripartire da una configurazione precedente. Prima di terminare la simulazione, infatti, vengono stampate sul file "config.final" le posizioni finali delle $108$ particelle. Se, in "input.dat", la variabile restart è pari ad uno, gli atomi di argon vengono disposti seguendo le indicazioni di "config.final". In caso contrario, vengono collocati sul reticolo fcc. Questa possibilità è stata ampiamente sfruttata nello studio del gas. Per raggiungere più velocemente l'equilibrazione, infatti, si è partiti dall'ultima disposizione delle particelle dello stato liquido. 

Non si è riusciti ad ottenere, per lo stato gassoso, una percentuale di mosse accettate del $50\%$. Nonostante si sia permesso uno spostamento quasi pari alla metà della dimensione del box di simulazione, l'accettazione rimaneva circa pari al $63\%$. Non avrebbe senso aumentare ulteriormente il range della distribuzione di probabilità di transizione $T(x|y)$. Con le periodic boundary conditions, infatti, tutti gli spostamenti superiori a $\frac{L}{2}$ riportano comunque l'atomo nel box. 

Questo risultato non deve sorprendere: in un gas la temperatura è relativamente alta e l'energia potenziale prossima a zero. Di conseguenza, $\exp(-\beta(E_{new}-E_{old}))\simeq1$ e, pertanto, molte mosse vengono accettate. 

Si studia, ora, l'andamento della funzione di autocorrelazione e le incertezze sulle grandezze termodinamiche al variare della dimensione $L$ dei blocchi. Si sono stampati, per ciascuno dei tre stati, $5\times10^5$ valori istantanei di pressione ed energia potenziale (per particella). Per definizione, per $\tau=0$, l'autocorrelazione dovrebbe essere pari ad uno, per poi decrescere, con un profilo approssimativamente esponenziale, all'aumentare di $\tau$. È logico attendersi una discesa molto più ripida per lo stato gassoso rispetto agli altri due. Il gas è, infatti, disordinato per definizione e dovrebbe, quindi, perdere memoria più velocemente delle sue precedenti configurazioni. Viceversa, nelle fasi condensate le disposizioni delle particelle risultano molto più simili tra loro, anche dopo diversi step. Nei grafici successivi $\tau\in[0;200]$.   

È necessario studiare l'andamento delle incertezze poiché, quando si hanno a disposizione dati tra loro correlati, se i blocchi non sono sufficientemente grandi, non è possibile applicare il teorema del limite centrale nelle sue ipotesi più semplici. Di conseguenza, con la canonica formula, si potrebbero ottenere delle sottostime di $\sigma$. Si è scelto di studiare il range $L\in[10;5000]$.

```python

def error(AV,AV2,n):  # Function for statistical uncertainty estimation
	if n==0:
		return 0
	else:
		return math.sqrt((AV2 - AV**2)/n)

def function(stato,grandezza):
	intestazione1="Stato "+stato
	intestazione2=grandezza
	print(intestazione1)
	print(intestazione2)
	if stato=="solido":
		titolo1="sol"
	if stato=="liquido":
		titolo1="liq"
	if stato=="gassoso":
		titolo1="gas"
	
	titolo="istantanei_"+titolo1+"_"+grandezza+".dat"
	
	energia = np.loadtxt(titolo, usecols=(0), delimiter=' ', unpack='true') 
	energia_2=energia**2

	media=np.sum(energia)/len(energia)
	sigma=np.sum(energia_2)/len(energia)
	sigma=sigma-media**2



	autocorr=np.zeros(200)


	for i in range(200):
		counter=0
		for j in range(len(energia)-i):
			autocorr[i]+=energia[j]*energia[i+j]
			counter=counter+1
		autocorr[i]=autocorr[i]/(counter)
		autocorr[i]=(autocorr[i]-media**2)/sigma



	x=np.arange(200)
	plt.plot(x,autocorr)
	plt.grid(True)
	plt.title("Andamento autocorrelazione")
	plt.xlabel(r"$\tau$")
	plt.ylabel("autocorrelazione")
	plt.show()

	L=[10,20,50,100,200,500,1000,2000,5000]

	for z in range(len(L)):
		
		N=int(len(energia)/L[z])
		ave=np.zeros(N)
		av2=np.zeros(N)
		for i in range(N):
			sum = 0
			for j in range(L[z]):
				k = j+i*L[z]
				sum += energia[k]
			ave[i] = sum/L[z]      
			av2[i] = (ave[i])**2 
	
		sum_prog=0
		su2_prog=0
		for i in range(N):
			sum_prog += ave[i] 
			su2_prog += av2[i]
		sum_prog/=(N) 
		su2_prog/=(N) 
		err_prog = error (sum_prog,su2_prog,N+1)
		plt.errorbar(L[z],sum_prog,yerr=err_prog,color="black",marker=".")
	
	plt.xscale('log')
	title="Andamento errori al variare della dimensione dei blocchi"
	plt.title(title)
	plt.xlabel("L")
	plt.ylabel(grandezza)
	plt.grid(True)
	plt.show()
```

```python
function("solido","energia")
function("solido","pressione")
function("liquido","energia")
function("liquido","pressione")
function("gassoso","energia")
function("gassoso","pressione")
```

Come previsto, l'autocorrelazione è pari ad uno per $\tau=0$ in tutti e sei i casi e decade più o meno velocemente. Ciò suggerisce che la funzione sia stata implementata correttamente. Nello stato gassoso la decrescita è molto più veloce: per $\tau\geq25$ l'autocorrelazione oscilla intorno a valori molto vicini allo zero. Si può notare, inoltre, che l'incertezza rimane all'incirca costante per dimensioni dei blocchi superiori a $50$.

Per le fasi condensate i comportamenti sono differenti: l'autocorrelazione cala molto più lentamente, specialmente per il liquido. Di conseguenza, è necessario utilizzare dei blocchi più grandi. Si osserva, infatti, che per lo stato liquido le $\sigma$ crescono fino a $L=500$, mentre, per quello solido, sembrano aumentare fino alla fine del range di $L$. Per questo motivo, neli esercizi successivi, le dimensioni dei blocchi verranno sempre poste pari a $10^4$.

Si può notare che, in alcuni dei grafici, al variare di $L$, a volte cambia pure il valore medio delle quantità. Tuttavia, è facile rendersi conto che si dovrebbero ottenere delle medie sempre identiche tra loro (se i blocchi possiedono tutti le stesse dimensioni). Questo problema è, quasi certamente, legato ad approssimazioni della macchina: osservando le scale dei grafici, ci si può accorgere che si sta lavorando con un numero piuttosto elevato di cifre decimali. Non è, pertanto, improbabile che sull'ultima si possa incappare in errori di arrotondamento.

Si studiano nel dettaglio, a questo punto, gli andamenti di $U/N$ e di $P$ per i tre stati di aggregazione. Dopo una fase iniziale di equilibrazione, analoga a quella precedentemente descritta, si eseguono simulazioni composte da $100$ blocchi.

Anche in questo caso i risultati (e le relative incertezze) sono espresse in unità di misura del sistema internazionale. Per convertire l'energia potenziale, la tecnica adottata è analoga a quella dell'esercitazione 4. Per la pressione, invece, si è invertita la semplice formula $P^*=\frac{P\sigma^3}{\epsilon}$.

Questi sono stati i risultati ottenuti:

```python
def grafici(stato):
	intestazione="Stato " + stato
	print(intestazione)
	
	if stato=="solido":
		titolo1="sol"
	if stato=="liquido":
		titolo1="liq"
	if stato=="gassoso":
		titolo1="gas"
		
	nome_file_1="output.epot."+titolo1+".0"
	
	bl,energia,err = np.loadtxt(nome_file_1, usecols=(0,2,3), delimiter=' ', unpack='true')
	
	energia=energia*120.*1.380649*(10**(-23)) #conversione in SI
	err=err*120.*1.380649*(10**(-23))
	
	plt.errorbar(bl,energia,yerr=err)
	plt.title("Andamento energia potenziale per particella")
	plt.xlabel("numero blocchi")
	plt.ylabel("U/N [J]")
	plt.grid(True)
	plt.show()
	
	nome_file_2="output.pres."+titolo1+".0"
	bl,pres,err = np.loadtxt(nome_file_2, usecols=(0,2,3), delimiter=' ', unpack='true')
	
	pres=pres*120.*1.380649*(10**(-23))/((0.34*(10**(-9)))**3)
	err=err*120.*1.380649*(10**(-23))/((0.34*(10**(-9)))**3)
	
	plt.errorbar(bl,pres,yerr=err)	
	plt.title("Andamento pressione")
	plt.xlabel("numero blocchi")
	plt.ylabel("P[Pa]")
	plt.grid(True)
	plt.show()	
```

```python
grafici("solido")
grafici("liquido")
grafici("gassoso")
```

Tutti i risultati sembrano piuttosto verosimili: l'energia potenziale è molto negativa per lo stato solido, in cui tutti gli atomi sono quasi fissati nelle loro posizioni di equilibrio. Per il gas, invece, è prossima a zero. In questo stato tutte le particelle si muovono pressoché liberamente.  

La pressione, invece, segue l'andamento opposto: è massima per il solido ed è minima per il gas. Si può cercare di giustificare anche questo risultato. Intuitivamente, infatti, più un materiale è denso (indipendentemente dal suo stato di aggregazione), maggiore dovrebbe essere la pressione che esercita (si pensi, ad esempio, alla legge dei gas perfetti, o alla legge di Stevino). In questa simulazione, in effetti, $\rho_{sol}^*=1,1$; $\rho_{liq}^*=0,8$; $\rho_{gas}^*=0,05$

Confrontando questi valori di $U/N$ con quelli ottenuti tramite l'algoritmo di dinamica molecolare, si può notare che non sono compatibili tra loro (entro le rispettive incertezze). Ciò non deve sorprendere o demoralizzare: in questo programma, infatti, sono state inserite anche le correzioni di coda. Le discrepanze sono riconducibili a questo contributo aggiuntivo precedentemente trascurato.

Si implementa ora il calcolo della distribuzione a due corpi $g(r)$. Si divide l'intervallo delle $r\in[0;\frac{L}{2}]$ in $100$ bin di dimensioni $dr$. Si valuta poi, in periodic boundary conditions, la distanza $d$ tra ogni coppia di particelle. Se $d$ appariene al bin j-esimo, si aumenta di due l'occorrenza di j. Si osservi che, poiché $[0;\frac{L}{2}]$ è stato diviso in parti uguali, vale semplicemente che $j=\lfloor \frac{d}{dr}\rfloor$. Infine, si normalizza l'istogramma moltiplicandolo per $\rho N \Delta V(r)$, con $\Delta V(r)=\frac{4\pi}{3}\left[(r+dr)^3-r^3\right]$. Anche in questo caso, per poter valutare le incertezze, si è utilizzata la media a blocchi. Nei file "output.gofr.0" si possono trovare, per ciascun blocco, i valori delle occorrenze dei bin. In "output.gave.0", invece, le occorrenze con le relative incertezze (calcolate considerando tutti i blocchi). Anche in questo caso si utilizzano $n_{blocks}=100$ e $L=10^4$.

La stessa implementazione viene anche eseguita nel programma di dinamica molecolare. Lo scopo, infatti, è quello di confrontare le $g(r)$ che si ottengono tramite algoritmi così diversi. Il codice dell'esercitazione 4 è stato poi semplificato il più possibile per renderlo veloce: non vengono più calcolate le varie quantità termodinamiche dal momento che si è solamente interessati alla $g(r)$. Si sono utilizzati, per le fasi condensate, un numero totale di step pari a $10^5$ (suddivisi in $100$ blocchi). Per il gas, invece, il valore non è sufficiente: si ottengono risultati non compatibili tra le varie simulazioni. Un problema simile era sorto durante lo studio della fase gassosa con l'algoritmo di Verlet. Pertanto, si è posto $n_{step}=10^6$ (anche in questo caso $n_{blocks}=100$).   

```python
print("Stato solido")
r,gr,err = np.loadtxt("output.gave.sol.0", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(r,gr,yerr=err, color ="red", label="Monte Carlo")

r,gr,err = np.loadtxt("output.gave_md.sol.0", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(r,gr,yerr=err, color ="blue", label="Dinamica molecolare")

plt.xlabel('$r^*$')
plt.ylabel('$g(r^*)$')
plt.grid(True)
legend = plt.legend(loc='best',shadow=True)
plt.show()

print("Stato liquido")
r,gr,err = np.loadtxt("output.gave.liq.0", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(r,gr,yerr=err, color ="red", label="Monte Carlo")

r,gr,err = np.loadtxt("output.gave_md.liq.0", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(r,gr,yerr=err, color ="blue", label="Dinamica molecolare")

plt.xlabel('$r^*$')
plt.ylabel('$g(r^*)$')
plt.grid(True)
legend = plt.legend(loc='best',shadow=True)
plt.show()

print("Stato gassoso")
r,gr,err = np.loadtxt("output.gave.gas.0", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(r,gr,yerr=err, color ="red", label="Monte Carlo")

r,gr,err = np.loadtxt("output.gave_md.gas.0", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(r,gr,yerr=err, color ="blue", label="Dinamica molecolare")

plt.xlabel('$r^*$')
plt.ylabel('$g(r^*)$')
plt.grid(True)
legend = plt.legend(loc='best',shadow=True)
plt.show()
```

Gli errori sono estremamente contenuti (sono quasi del tutto invisibili) per entrambi gli algoritmi. Si può notare come, per tutti e tre gli stati, le $g(r)$ risultino perfettamente compatibili tra loro, nonostante siano state ricavate con algoritmi totalmente diversi. Ciò fornisce un'ulteriore, definitiva, prova della potenza pressoché illimitata degli algoritmi Monte Carlo. In questo caso, infatti, si ottengono delle distribuzioni perfettamente identiche a quelle che si ricavano seguendo le traiettorie delle particelle nel tempo.
