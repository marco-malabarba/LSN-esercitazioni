### **INTEGRAZIONE MONTE CARLO**

### Calcolare $I = \int_0^1 \frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right) dx = 1$
1. Con una distribuzione di probabilità $p(x)$ uniforme su $[0;1]$
2. Con una distribuzione di probabilità non uniforme



Un integrale della forma: $I=\int_a^b g(x)p(x)dx$, con p(x) generica distribuzione di probabilità su $[a;b]$, si può facilmente calcolare mediante tecniche Monte Carlo. Dato un insieme di variabili $x_1,x_2,\dots,x_n$, distribuite secondo $p(x)$, si può definire una quantità $G_N=\frac{1}{N}\sum_{i=1}^N g(x_i)$. Se l'integrale $I$ è finito, per $N\to+\infty$, $\langle G_N \rangle=I$.

In questo caso, si può considerare, inizialmente, una $p(x)=1$, quindi uniforme sull'intervallo di integrazione. È evidente che $p(x)$ sia ben definita e normalizzata su $[0;1]$.

Si vuole graficare, per un numero crescente di passi Monte Carlo, la stima dell'integrale con la relativa incertezza. Viene dunque utilizzato il metodo a blocchi descritto nell'esercitazione precedente. 

Si generano $M=10^6$ quindi numeri casuali uniformemente distribuiti tra $[0;1]$ e si divide l'esperimento in $N=100$ blocchi. Ciascuno di essi consiste di $L=\frac{M}{N}=10^4$ valori casuali $x$ con i quali, mediante la formula precedente, si otterranno $100$ stime di $I$. Per poter disegnare il grafico richiesto è quindi sufficiente considerare un numero crescente di blocchi: inizialmente uno solo, poi due, tre, fino ad arrivare a cento.

Il numero di passi Monte Carlo è dato semplicemente da $LN_c$. La stima dell'integrale è data, invece, dalla media degli $N_c$ blocchi considerati. Gli errori, infine, si ricavano facilmente dalle equazioni esposte nell'esercitazione precedente.

Questi sono stati i risultati ottenuti:


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

i, intg, err = np.loadtxt("integrale1.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,intg,yerr=err)
plt.title('Calcolo integrale')
plt.xlabel('numero passi')
plt.ylabel('I')
plt.grid(True)
N = 100
x = np.linspace(0.,1000000.,N,endpoint=True)
y=np.zeros(N)
y=y+1.
plt.plot(x,y,color="red") 
plt.show()
```

Il valore di $I$ può anche calcolare mediante una semplice integrazione analitica e risulta $I=1$. La linea orizzontale rossa sul grafico aiuta a visualizzare l'andamento del calcolo Monte Carlo e la sua convergenza verso questo valore "atteso". 

Si può dunque osservare che le quantità ottenute numericamente risultano sempre compatibili, entro $2 \sigma$, con $I=1$ . Inoltre, i valori finali (che si ottengono mediando quasi tutti i cento blocchi) sono compatibili, entro le loro incertezze, con la previsione teorica. Ci si può dunque ritenere soddisfatti dei risultati di questa prima simulazione.

Per gli ultimi dati, l'incertezza è pari a circa $5\times10^{-4}$. L'obiettivo della prossima fase dell'esperimento è cercare di ridurre questo valore. 

Una soluzione immediata per ottenere risultati più precisi è aumentare il valore di $M$ tenendo fisso, al contempo, $N$. In questo modo, infatti, ciascun blocco sarebbe costituito da un maggior numero di passi e, di conseguenza, diminuirebbero le fluttuazioni statistiche tra i vari $I_i$. Tuttavia questo metodo non è particolarmente raffinato: aumentando $M$ cresce anche il tempo computazionale. 

Si può, invece, utilizzare una distribuzione di probabilità $p(x)$ diversa (tecnica dell'**importance sampling**). Per poter continuare a valutare lo stesso $I$, l'integranda precedente $g(x)=\frac{\pi}{2}\cos(\frac{\pi}{2}x)$ diviene ora $\bar{g}(x)=\frac{g(x)}{p(x)}$. Da questa formula si comprende che la $p(x)$ scelta, oltre ad essere non negativa, deve anche essere diversa da zero su tutto l'intervallo $[0;1]$. 

In generale la varianza sul calcolo dell'integrale è pari a: $\sigma^2=\langle g^2 \rangle - \langle g \rangle^2$. Si può dimostrare la varianza diminuisce se la $p(x)$ è simile a $g(x)$.

$p(x)=1$ è il polinomio di Taylor di grado zero di $\cos(\frac{\pi}{2}x)$ centrato in $x=0$. Per ottenere una migliore "somiglianza" si può procedere nello sviluppo. Al secondo ordine risulta: $1-\frac{\pi^2x^2}{8}$. Si potrebbe dunque scegliere questa $p(x)$ (con un opportuno fattore di normalizzazione). Tuttavia, risulta evidente che questa candidata non soddisfa alla richiesta di non-negatività presente nella definizione di distribuzione di probabilità. 

Aggiungendo un ulteriore termine allo sviluppo si ottiene $1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}$. Gli zeri di questa funzione sono tutti esterni all'intervallo $[0;1]$, pertanto $$p(x)=\frac{1920}{\pi^4-80\pi^2+1920}\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right)$$ Il prefattore è stato inserito in modo tale che risulti: $\int_0^1p(x)dx=1$. Di conseguenza, $\bar{g}(x)=\frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right)\frac{\pi^4-80\pi^2+1920}{1920\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right)}$. 

È dunque necessario generare dei numeri casuali distribuiti secondo la nuova $p(x)$. Non è possibile utilizzare il metodo dell'inversione della funzione cumulativa: non si è in grado di risolvere analiticamente un'equazione di quinto grado. Si decide perciò di applicare tecniche di **rigetto**. Vengono generati due numeri casuali $x$ ed $r$ con distribuzione uniforme. $r \in [0;1]$, mentre $x \in [a,b]$ in cui $a$ e $b$ sono gli estremi di definizione della $p(x)$ (in questo caso sono anch'essi con zero e uno). Successivamente si calcola il massimo $p_{max}$ della distribuzione, e, in questo caso, risulta $p_{max}=\frac{1920}{\pi^4-80\pi^2+1920}$. Se $r<\frac{p(x)}{p_{max}}$, si accetta il valore di $x$ precedentemente generato, altrimenti lo si rigetta. I valori che soddisfano alla disequazione precedente sono distribuiti secondo la $p(x)$ richiesta. 

Si ripete, dunque, l'esperimento precedentemente descritto con gli stessi valori di $M$ ed $N$. Dal momento che si sta utilizzando una tecnica di rigetto, per effettuare un passo Monte Carlo non è sempre sufficiente estrarre un singolo numero casuale $x\in[0;1]$: il singolo passo non può dirsi concluso finché $x$ non risulta accettabile. 

Questi sono stati i risultati ottenuti:

```python
i, intg, err = np.loadtxt("integrale2.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,intg,yerr=err)
plt.title('Calcolo integrale')
plt.xlabel('numero passi')
plt.ylabel('I')
plt.grid(True)
plt.plot(x,y,color="red") 
plt.show()
```

Come si può notare, i risultati ottenuti sono, anche in questo caso, compatibili con il valore atteso. Tuttavia le incertezze ottenute sono estremamente più piccole: per gli ultimi dati sono, approssimativamente, $2,6\times10^{-5}$. 

Il rapporto tra le incertezze delle due simulazioni vale circa $19$. Questo valore è in accordo con la formula generale della varianza precedentemente introdotta. $\langle g \rangle^2=1$ per entrambi gli esperimenti, mentre $\int_0^1\left(\frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right)\right)^2dx=\frac{\pi^2}{8}\simeq1,233$ e $\int_0^1\left(\frac{\pi}{2}\cos(\frac{\pi}{2}x)\frac{\pi^4-80\pi^2+1920}{1920\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right)}\right)^2\frac{1920}{\pi^4-80\pi^2+1920}\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right) dx \simeq1,000672$. Il rapporto tra le varianze è circa $350$, da cui, estraendo la radice, si ricava che il rapporto atteso tra le $\sigma$ è $18,6$.

Da queste semplici considerazioni si potrebbe assumere che il secondo metodo sia dunque estremamente migliore del primo, dal momento che si ottengono risultati venti volte più precisi. Tuttavia bisogna tenere in considerazione anche il tempo computazionale $t$. Utilizzando tecniche di rigetto, infatti, $t$ non può che aumentare: i numeri casuali effettivamente generati possono essere ben più di $M=10^6$. 

Nel calcolo di integrali con tecniche Monte Carlo, si utilizza spesso un parametro che misura l'efficienza dell'algoritmo: $Q_i=t_i\sigma_i^2$. Un metodo è tanto più efficiente quanto più è piccolo il suo $Q_i$. Per confrontare correttamente i due diversi metodi utilizzati, si dovrebbe stimare anche il rapporto tra i tempi computazionali. 

Il suo ordine di grandezza si può calcolare, in modo molto grossolano, dal rapporto dei seguenti integrali:

L'interpretazione geometrica del metodo del rigetto è la seguente: la coppia $(x;r)$ viene accettata se il punto $(x;p_{max}r)$ si trova nella parte di piano compresa tra l'asse delle ascisse e la $p(x)$. La probabilità che ciò accada è dunque data dal rapporto tra gli integrali di $p(x)$ e di $p_{max}$ (calcolati ovviamente su $[0;1]$). In media, ogni volta, occorre generare $x$ un numero di volte pari all'inverso di questo rapporto (è il valore che si può osservare nel disegno). Il tempo computazionale risulta quindi aumentato di circa un fattore $1,6$.

Questa è soltanto una prima stima di $t_i$: volendo essere più precisi bisognerebbe tenere conto che, per applicare la tecnica del rigetto, servono due numeri casuali (anche $r$ oltre ad $x$). $t$ andrebbe quindi moltiplicato per un ulteriore fattore due. Inoltre, anche per calcolare le varie $g(x_i)$ i tempi impiegati potrebbero non essere esattamente gli stessi (specie se la $\bar g$ è molto più complessa di $g$). 

In ogni caso, si può certamente affermare che questo secondo metodo sia il più efficiente: la varianza è stata ridotta di circa $350$ volte, mentre il tempo computazionale è dello stesso ordine di grandezza della prima simulazione. 

Si ripete l'esperimento una terza volta, cercando però di utilizzare una $p(x)$ che possa essere invertita analiticamente. Eliminando le complicazioni del rigetto, è infatti possibile confrontare direttamente le efficienze dei metodi tramite il rapporto delle varianze. 

Non si ha una vasta scelta per $p(x)$: se fosse un polinomio, dovrebbe essere necessariamente di primo grado, dal momento che non si sanno risolvere esattamente equazioni di grado superiore al secondo. Si decide quindi di sviluppare $\cos(\frac{\pi}{2}x)$ al primo ordine in $x=\frac{1}{2}$. Con la corretta normalizzazione risulta che:
$$p(x)=1-\frac{\pi}{2}\left(x-\frac{1}{2}\right) \quad \quad \bar{g}(x)=\frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right)\frac{1}{1-\frac{\pi}{2}\left(x-\frac{1}{2}\right)}$$

La funzione cumulativa $F(x)=-\frac{\pi x^2}{4}+\frac{\pi x}{4}+x$ è ora facilmente invertibile: si ottiene che $$x=\frac{\pi+4-\sqrt{(\pi+4)^2-16\pi y}}{2\pi}$$ è distribuito secondo $p(x)$ ($y$ è un numero casuale uniformemente distribuito tra $[0;1]$). Delle due soluzioni dell'equazione di secondo grado si è considerata solo quella che restituisce, per ogni $y$, un valore di $x$ nell'intervallo desiderato.  

In questa fase dell'esperimento non vengono modificati i valori di $M$ e di $N$ e, ripetendo tutte le operazioni precedentemente descritte, si ottengono i seguenti risultati:

```python
i, intg, err = np.loadtxt("integrale3.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,intg,yerr=err)
plt.title('Calcolo integrale')
plt.xlabel('numero passi')
plt.ylabel('I')
plt.grid(True)
plt.plot(x,y,color="red") 
plt.show()
```

Anche in questo caso, i risultati sono sempre compatibili con il valore atteso dell'integrale entro $2\sigma$. Il valore ottenuto con tutti i cento blocchi è, tuttavia, leggermente più piccolo di uno, pur considerando la sua incertezza. Ciò non deve sorprendere eccessivamente: la distanza è estremamente piccola ed è attribuibile a fluttuazioni statistiche. Bisogna sempre ricordare, inoltre, che, se il risultato $x$ di una misura è distribuito secondo una curva gaussiana definita dal valore medio $\langle X \rangle$ e dalla deviazione standard $\sigma$, vi è una probabilità di circa $\frac{1}{3}$ che $x \notin [\langle X \rangle-\sigma;\langle X \rangle+\sigma]$ 

In questo caso $\langle g^2 \rangle - \langle g \rangle^2\simeq0,01785$, pertanto gli errori dovrebbero risultare $3,6$ volte più piccoli di quelli ottenuti tramite la distribuzione uniforme. Poiché gli ultimi dati del grafico presentano un'incertezza di circa $1,3\times10^{-4}$, Il rapporto tra queste $\sigma$ e quelle primo esperimento è di circa $3,8$. Anche in questo caso, quindi, il valore è molto simile a quello atteso.

Questi risultati, sono, invece, molto meno precisi di quelli ottenuti dal secondo esperimento. In questo caso, per poter invertire la cumulativa, ci si è dovuti accontentare di un'approssimazione lineare della funzione. Sicuramente questa $p(x)$ "somiglia" di meno a $g(x)$ rispetto a quella precedente ricavata da uno sviluppo al quarto ordine.

Si è dunque riusciti anche a verificare quantitativamente che, più la distribuzione di probabilità è simile all'integranda, minore sarà la varianza nel calcolo di $I$.

### **RANDOM WALK TRIDIMENSIONALE**
### Simulare un random walk e disegnare, in funzione del numero di step $i\in [0;100]$, $\sqrt{\langle |\vec{r
}(i)|^2 \rangle}$ con le relative incertezze, nei due seguenti casi:
1. **RANDOM WALK SU UN RETICOLO CUBICO**, ad ogni step la particella può muoversi in avanti o indietro in **una** delle tre direzioni principali (si assume che ciascuna abbia la stessa probabilità)

2. **RANDOM WALK NEL CONTINUO**, ad ogni step la particella può muoversi in una generica direzione dello spazio (si assume che ognuna abbia la stessa probabilità)
 


Si assume che la particella parta dall'origine del sistema cartesiano e che, ad ogni step, si sposti di una quantità fissata $a=1$. Per poter ottenere risultati statisticamente significativi, si devono simulare parecchi random walk (in questo caso si è scelto $M=10^4$). 

Fissando inizialmente il numero $i$ di step "percorsi", per poter analizzare i risultati dell'esperimento si utilizzano tre array (x,y,z) di dimensione $M$. $(x[k];y[k];z[k])$ rappresenta la posizione nello spazio della particella durante la k-esima simulazione di random walk ($k=1,2,\dots,M$). Per ogni $M$ si può dunque calcolare $|\vec{r_j}|^2 $ ($j=1,2,\dots,M$) tramite il teorema di Pitagora.

Per poter calcolare le incertezze si divide, ancora una volta, l'esperimento in $N=100$ blocchi. Ciascuno di essi contiene, quindi, $L=\frac{M}{N}$ simulazioni di random walk. Si possono ottenere $N$ valori di $\langle |\vec{r_k}|^2 \rangle$ ($k=1,2,\dots,N$) mediando gli $L$ $ |\vec{r_j}|^2 $ presenti in ogni blocco. 

Come negli esercizi precedenti il valore atteso di $\langle |\vec{r}|^2 \rangle$ è la media della delle cento stime e la sua incertezza si può determinare mediante l'usuale formula.

Poiché si vuole graficare $\sqrt{\langle |\vec{r}|^2 \rangle}$, è sufficiente estrarre la radice e, per la $\sigma$, utilizzare la teoria di propagazione degli errori. Se una grandezza è funzione di un'altra (di cui è nota la deviazione standard) $y=y(x)$, vale che $\sigma_y=\left\lvert{\frac{dy(x)}{dx}}\right\lvert\sigma_x$. In questo caso $\sigma_r=\frac{1}{2\sqrt{\langle |\vec{r}|^2 \rangle}}\sigma_{r^2}$

Si ripete la stessa analisi per tutti gli step $i=1,2,\dots,100$ che si desidera analizzare. Inizialmente (dopo zero step) tutte le componenti degli array sono pari a zero (si parte sempre dall'origine). Bisogna implementare, a questo punto, un algoritmo che permetta di aggiornare le componenti dei tre array a seguito di uno step. 

Per il random walk sul reticolo si genera un numero casuale (estratto da una distribuzione uniforme) $v\in[0;6)$ che determina la direzione (e il verso) dello step. Si è adottata la seguente convenzione: se $\lfloor{v}\rfloor=0$, lo spostamento avviene di $a$ lungo l'asse x; se $\lfloor{v}\rfloor=1$, lo spostamento avviene nel verso negativo di x. Si procede (ordinatamente) con la stessa logica per gli altri quattro valori possibili di $\lfloor{v}\rfloor$. Il metodo viene ripetuto per ciascuno degli $M$ random walk che si desidera simulare.

Questa scelta è, ovviamente, del tutto arbitraria: si potrebbe associare in modo diverso la direzione dello spostamento alla parte intera del numero casuale (ad esempio partendo dall'asse z). Tuttavia, poiché si utilizza una distribuzione uniforme, i risultati finali non cambiano, specialmente quando, nell'intera simulazione, è necessario generare moltissimi $v$.

Per il random walk nel continuo, invece, è sufficiente determinare le distribuzioni di $\theta$ e $\phi$ che campionino uniformemente l'angolo solido. Si può dimostrare che è sufficiente generare $\phi\in[0;2\pi)$ uniformemente e $\theta=\arccos{(1-2r)}$. $r$ è estratto da una densità di probabilità costante su $[0;1)$. Tramite la coppia $(\theta;\phi)$ e le equazioni delle coordinate sferiche, si possono aggiornare le tre componenti della posizione della particella. Anche in questo caso si ripete il procedimento $10^4$ volte.

Dopo aver disegnato i due grafici, si effettua un fit con una funzione del tipo $f(i)=k\sqrt i$, con $i$ numero di step percorsi. Ci si aspetta che $k=a$ in entrambi i casi. Vale infatti che: $$\langle |\vec{r}(i)|^2 \rangle=\langle \left(\vec{r_1}+\vec{r_2}+\dots+\vec{r_i}\right)^2\rangle=\langle|\vec{r_1}|^2\rangle+\langle|\vec{r_2}|^2\rangle+\dots\langle|\vec{r_i}|^2\rangle+2\left(\langle\vec{r_1}\vec{r_2}\rangle+\langle\vec{r_1}\vec{r_3}\rangle+\dots+\langle\vec{r_1}\vec{r_i}\rangle+\langle\vec{r_2}\vec{r_3}\rangle+\dots\right)$$

Nella formula precedente si è scomposto il cammino complessivo nella somma (vettoriale) degli $i$ step. Nell'ultimo membro la si è riscritta in una maniera più conveniente. Gli $\langle|\vec{r_j}|^2\rangle=a^2$ per definizione, inoltre, nel random walk sul reticolo, è evidente che tutti i valori medi $\langle\vec{r_j}\vec{r_k}\rangle=0$. Pertanto  $\langle |\vec{r}(i)|^2 \rangle=ia^2$

Per il random walk nel continuo la dimostrazione è leggermente più complicata e può essere effettuata per induzione. Per un cammino di un singolo step vale, chiaramente, $\langle |\vec{r}(1)|^2 \rangle=a^2$. Si suppone ora che la relazione $\langle |\vec{r}(i)|^2 \rangle=ia^2$ sia valida per un cammino di n step. Al passo successivo, scomponendo in un modo simile al precedente, si ha che: $\langle |\vec{r}(n+1)|^2 \rangle=\langle \left(\vec{r}(n)+\vec{a}\right)^2 \rangle=\langle |\vec{r}(n)|^2 \rangle+a^2+\langle 2ar(n)\cos\theta\rangle$.

Nell'ultima relazione il primo addendo è pari a $na^2$ per l'ipotesi induttiva, mentre l'ultimo si può calcolare con un semplice integrale poiché è nota la distribuzione di probabilità sull'angolo solido. $\int{d\Omega 2ra\cos\theta\frac{1}{4\pi}}=0$. Quindi, anche in questo caso, si è verificata la relazione $\langle |\vec{r}(i)|^2 \rangle=ia^2$.
Dopo aver simulato i due esperimenti si sono ottenuti i seguenti risultati: 

```python
from scipy.optimize import curve_fit

def f(x,k):  
    return k*np.sqrt(x)

i, r, err = np.loadtxt("RWdiscreto.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,r,yerr=err)
plt.title("Random walk discreto")
plt.xlabel('numero step')
plt.ylabel('$\sqrt{<|r_{step}|^2>}$')
plt.grid(True)
p_opt, p_cov = curve_fit(f,i,r)
N = 100
x = np.linspace(1.,100.,N,endpoint=True)
y_fit = f(x,p_opt[0])

plt.plot(x,y_fit,color="red") 
plt.show()


print("Parametro k di best-fit:")
print(p_opt)

print("Incertezza sul parametro=")
print(np.sqrt(np.diagonal((p_cov))))

E_i=f(x,1) #calcolo chi-quadro con il parametro atteso e non di best-fit!
sigma=err
chiquadro_i=np.zeros(np.size(x))
for i in range(np.size(x)):
	if sigma[i]==0: #dopo 1 step il dato ha incertezza zero (non lo uso nel calcolo) chi-quadro
		chiquadro_i[i]=0
	else: 
		chiquadro_i[i]=((E_i[i]-r[i])/sigma[i])**2
chiquadro=np.sum(chiquadro_i)
print('Chi-quadro:')
print(chiquadro)
```

```python
i, r, err = np.loadtxt("RWcontinuo.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,r,yerr=err)
plt.title('Random walk continuo')
plt.xlabel('numero step')
plt.ylabel('$\sqrt{<|r_{step}|^2>}$')
plt.grid(True)
p_opt, p_cov = curve_fit(f,i,r)
plt.plot(x,y_fit,color="red") 
plt.show()
print("Parametro k di best-fit:")
print(p_opt)

print("Incertezza sul parametro=")
print(np.sqrt(np.diagonal((p_cov))))

E_i=f(x,1) #calcolo chi-quadro con il parametro atteso e non di best-fit!
sigma=err
chiquadro_i=np.zeros(np.size(x))
for i in range(np.size(x)):
	if sigma[i]==0: 
		chiquadro_i[i]=0
	else: 
		chiquadro_i[i]=((E_i[i]-r[i])/sigma[i])**2
chiquadro=np.sum(chiquadro_i)
print('Chi-quadro:')
print(chiquadro)
```

Come si può immediatamente notare, le incertezze sono talmente piccole da risultare pressoché invisibili: nessun valore di $\sqrt{\langle |\vec{r}(i)|^2 \rangle}$ ha un errore superiore a $0,045$. Le due funzioni di best-fit (in rosso) sembrano, in ogni caso, compatibili con i risultati ottenuti: sono quasi totalmente sovrapposte con la curve definite dai dati sperimentali (in blu).

Si nota che, dopo un singolo step, si ottiene un valore di $\sqrt{\langle |\vec{r}(1)|^2 \rangle}$ privo di incertezze per entrambi i random walk. Infatti, se $i=1$, si ha necessariamente che $\sqrt{\langle |\vec{r}|^2 \rangle}=a$ per tutte le simulazioni (può cambiare solo la direzione del moto ma non la lunghezza del "passo"). La deviazione standard di un insieme di valori identici è, ovviamente, zero. 

I due valori di $k$ di best-fit potrebbero sembrare molto vicini ad $a$: la differenza percentuale con è, appena, di qualche per mille. Tuttavia è facile osservare che non sono compatibili con questo valore previsto entro le incertezze: la loro distanza è di decine di $\sigma$. 

Questo risultato è piuttosto sorprendente, si è deciso allora di effettuare un test del $\chi^2$ utilizzando, come valori attesi, $f(i)=a\sqrt i$. In questo modo si può comprendere se è possibile rigettare l'ipotesi che $k=a$ e, in caso affermativo, con quale livello di confidenza. A differenza dell'esercitazione precedente, non è più lecito approssimare la varianza con il valore atteso: ciascun $\sqrt{\langle |\vec{r}(i)|^2 \rangle}$ possiede una propria $\sigma$ che deve essere inserita nella definizione generale del $\chi^2$.

In entrambi i casi si ottiene un valore significativamente inferiore al numero di gradi di libertà del sistema: di conseguenza, poiché il $\chi^2_{rid}<1$, non è possibile rigettare l'ipotesi che $k=a$.  

Bisogna sempre ricordare che esistono diversi metodi per trovare i parametri (e le incertezze) di una curva di best-fit. Ciascuno di essi produce dei risultati che possono essere diversi dagli altri. Come si è spiegato, infatti, non è possibile rigettare l'ipotesi che $k=1$ con i dati raccolti dalle simulazioni (nonostante questo valore si trovi a diverse $\sigma$ dal $k_{best}$). 

In ogni caso, è lecito attendersi che, all'aumentare del numero dei random walk, i risultati convergano verso il valore $k=a$. Se $M\to+\infty$, infatti, ciascun valore di $\sqrt{\langle |\vec{r}(i)|^2 \rangle}\to a\sqrt{i}$. Le fluttuazioni statistiche di ogni dato dovrebbero quindi diminuire e, di conseguenza, qualunque fit produrrebbe risultati più in linea con i modelli teorici.

Per verificare questa supposizione si simulano altri $M=10^5$ random walk nel continuo: ci si aspetta che $k$ sia più vicino (in unità $\sigma$) ad $1$. 

```python
i, r, err = np.loadtxt("RWcontinuo2.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,r,yerr=err)
plt.title('Random walk continuo')
plt.xlabel('numero step')
plt.ylabel('$\sqrt{<|r_{step}|^2>}$')
plt.grid(True)
p_opt, p_cov = curve_fit(f,i,r)
plt.plot(x,y_fit,color="red") 
plt.show()
print("Parametro k di best-fit:")
print(p_opt)

print("Incertezza sul parametro=")
print(np.sqrt(np.diagonal((p_cov))))

E_i=f(x,1) #calcolo chi-quadro con il parametro atteso e non di best-fit!
sigma=err
chiquadro_i=np.zeros(np.size(x))
for i in range(np.size(x)):
	if sigma[i]==0: 
		chiquadro_i[i]=0
	else: 
		chiquadro_i[i]=((E_i[i]-r[i])/sigma[i])**2
chiquadro=np.sum(chiquadro_i)
print('Chi-quadro:')
print(chiquadro)
```

La previsione è confermata dai risultati: la distanza tra il valore atteso e quello ottenuto è ora inferiore a $2\sigma$. Anche in questo caso, inoltre, si osserva un $\chi^2_{rid}<1$ e non è quindi possibile rigettare l'ipotesi che i dati obbediscano a una relazione del tipo $\sqrt{\langle |\vec{r}(i)|^2 \rangle} =a\sqrt i$.

Tutti questi risultati suggeriscono un comportamento diffusivo. Risolvendo l'equazione di diffusione con una funzione di distribuzione a delta di Dirac nell'origine per $t=0$, si ottiene che $\langle |\vec{r}(t)|^2 \rangle \propto t$. La legge di diffusione è una un'equazione differenziale alle derivate parziali, quindi occorre che gli step siano infinitesimi e il tempo scorra in modo continuo. Nelle simulazioni effettuate la particella è sempre partita dall'origine, ma, per semplicità, si è assunto che gli step avessero una lunghezza finita ($a=1$). È lecito supporre, inoltre, che il tempo necessario per percorrere ciascuno step sia sempre lo stesso ($\tau$) e, di conseguenza, $t=i\tau$. Sostituendo $i$ nella formula $\langle |\vec{r}(i)|^2 \rangle=ia^2$,  si ricava che, anche per questi random walk "a passi discreti", $\langle |\vec{r}(t)|^2 \rangle \propto t$

Questa è una relazione di proporzionalità approssimata: in linea di principio varrebbe solamente per tempi multipli interi di $\tau$: non si definisce una $\langle |\vec{r}|^2\rangle$ nel mezzo di uno step discreto. Se si considerassero scale temporali con tempi caratteristici $t>>\tau$ l'approssimazione diventerebbe sempre più precisa. Per $\tau\to0$, anche $a\to0$ (poiché nessuna particella può muoversi a velocità infinita), e, come ci si poteva attendere, si torna nel limite del continuo in cui la proporzionalità $\langle |\vec{r}(t)|^2 \rangle \propto t$ è esatta.
