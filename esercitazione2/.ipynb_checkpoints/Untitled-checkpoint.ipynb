### **INTEGRAZIONE MONTE CARLO**

### Calcolare $I = \int_0^1 \frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right) dx = 1$
1. Con una distribuzione di probabilità $p(x)$ uniforme su $[0;1]$
2. Con una distribuzione di probabilità non uniforme



- Un generico integrale della forma: $I=\int_a^b g(x)p(x)dx$, con p(x) generica distribuzione di probabilità su $[a;b]$, si può facilmente calcolare mediante tecniche Monte Carlo. Dato un insieme di variabili $x_1,x_2,\dots,x_n$, distribuite secondo $p(x)$, si può definire una quantità $G_N=\frac{1}{N}\sum_{i=1}^N g(x_i)$. Se l'integrale $I$ è finito, per $N\to+\infty$, $\langle G_N \rangle=I$.
- In questo caso, si può considerare, inizialmente, una $p(x)=1$, quindi uniforme sull'intervallo di integrazione. È evidente che questa $p(x)$ sia ben definita e normalizzata sull'intervallo $[0;1]$.
- Si vuole graficare, per un numero crescente di passi Monte Carlo, la stima dell'integrale con la relativa incertezza. Viene dunque utilizzato il modello a blocchi descritto nell'esercitazione precedente. 
- Vengono dunque generati $M=10^6$ numeri casuali uniformemente distribuiti tra $[0;1]$ e si divide l'esperimento in $N=100$ blocchi. Ciascun blocco consiste di $L=\frac{M}{N}=10^4$ valori di $x$; mediante la formula precedente, si otterranno, dunque, $100$ stime di $I$. Per poter disegnare il grafico richiesto è quindi sufficiente considerare un numero crescente di blocchi: inizialmente uno solo, poi due, tre, fino ad arrivare a cento.
-  Il numero di passi Monte Carlo è dato semplicemente da $LN_c$. La stima dell'integrale è data, invece, dalla media degli $N_c$ blocchi considerati. La relativa incertezza, infine, si ricava facilmente dalle formule esposte nell'esercitazione precedente.
- Questi sono stati i risultati ottenuti:


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

i, intg, err = np.loadtxt("integrale1.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,intg,yerr=err)
plt.xlabel('numero passi')
plt.ylabel('I')
plt.grid(True)
N = 100
x = np.linspace(0.,1000000.,N,endpoint=True)
y=np.zeros(N)
y=y+1.
plt.plot(x,y,color="red") 
plt.show()
```

- Il valore di $I$ può calcolare mediante una semplice integrazione analitica e risulta, evidentemente, $I=1$. La linea orizzontale rossa sul grafico aiuta a visualizzare l'andamento del calcolo Monte Carlo e la sua convergenza verso il valore "vero". 
- Si può dunque osservare che il valore ottenuto numericamente risulta sempre compatibile, entro $2 \sigma$, con quanto atteso teoricamente. Inoltre, i valori finali (che si ottengono considerando quasi tutti i cento blocchi) risultano essere compatibili, entro le loro incertezze, con $I=1$. Ci si può dunque ritenere soddisfatti dei risultati di questa prima simulazione.
- Per gli ultimi dati, l'incertezza risulta pari a circa $5\times10^{-4}$. L'obiettivo della prossima fase dell'esperimento è cercare di ridurre questo valore. 
- Un modo immediato per ottenere risultati più precisi è aumentare il valore di $M$ tenendo fisso, al contempo, $N$. In questo modo, infatti, ciascun blocco sarebbe costituito da un maggior numero di passi e, di conseguenza, diminuirebbero le fluttuazioni statistiche tra i vari $I_i$. Tuttavia questo metodo non è particolarmente raffinato: aumentando $M$ cresce anche il tempo computazionale. 
- Si può, invece, utilizzare una distribuzione di probabilità $p(x)$ diversa (tecnica dell'**importance sampling**). Per poter continuare a valutare lo stesso $I$, l'integranda precedente $g(x)=\frac{\pi}{2}\cos(\frac{\pi}{2}x)$ diventerà ora $\bar{g}(x)=\frac{g(x)}{p(x)}$. Da questa formula si comprende che la $p(x)$ scelta, oltre a dover essere non negativa, deve anche essere diversa da zero su tutto l'intervallo $[0;1]$. 
- In generale la varianza sul calcolo dell'integrale è pari a: $\langle g^2 \rangle - \langle g \rangle^2$. Si può dimostrare la varianza diminuisce se la $p(x)$ è simile a $g(x)$.
- La $p(x)=1$ precedente è lo sviluppo di Taylor all'ordine zero di $\cos(\frac{\pi}{2}x)$ centrato in $x=0$. Per ottenere una migliore "somiglianza" si può procedere nello sviluppo. Al secondo ordine risulta: $1-\frac{\pi^2x^2}{8}$. Si potrebbe dunque scegliere questa $p(x)$ (con un opportuno fattore di normalizzazione). Tuttavia, risulta evidente che questa candidata non soddisfa la richiesta di non-negatività della definizione di distribuzione di probabilità. 
- Aggiungendo un ulteriore termine allo sviluppo si ottiene $1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}$. Gli zeri di questa funzione sono tutti esterni all'intervallo $[0;1]$, pertanto $$p(x)=\frac{1920}{\pi^4-80\pi^2+1920}\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right)$$ Il prefattore è stato inserito in modo tale che risulti: $\int_0^1p(x)dx=1$. Di conseguenza, $\bar{g}(x)=\frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right)\frac{\pi^4-80\pi^2+1920}{1920\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right)}$. 
- Serve dunque generare dei numeri casuali distribuiti secondo la distribuzione $p(x)$. Non è possibile utilizzare il metodo dell'inverso della funzione cumulativa: non si è in grado di risolvere analiticamente un'equazione di quinto grado. Si decide perciò di applicare tecniche di **rigetto**. Vengono generati due numeri casuali $x$ ed $r$ con distribuzione uniforme. $r \in [0;1]$, mentre $x \in [a,b]$ in cui $a$ e $b$ sono gli estremi di definizione della $p(x)$ (in questo caso coincidono anch'essi con zero e uno). Successivamente si calcola il massimo $p_{max}$ della distribuzione, in questo caso risulta ovviamente $p_{max}=\frac{1920}{\pi^4-80\pi^2+1920}$. Se $r<\frac{p(x)}{p_{max}}$, si accetta il valore di x precedentemente generato, altrimenti lo si rigetta. I valori accettati sono distribuiti secondo la $p(x)$ richiesta. 
- Si ripete, dunque, l'esperimento precedentemente descritto con gli stessi valori di $M$ ed $N$. Dal momento che si è modificata la $p(x)$ (e quindi di conseguenza $g(x) \to \bar{g}(x)$), per effettuare un passo Monte Carlo non sempre è sufficiente generare un singolo numero casuale $x\in[0;1]$. Poiché si sta usando una tecnica di rigetto il singolo passo non può dirsi concluso finché $x$ non soddisfa il criterio necessario per essere accettato. 
- Questi sono stati i risultati ottenuti:

```python
i, intg, err = np.loadtxt("integrale2.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,intg,yerr=err)
plt.xlabel('numero passi')
plt.ylabel('I')
plt.grid(True)
plt.plot(x,y,color="red") 
plt.show()
```

- Come si può notare, i risultati ottenuti sono, anche in questo caso, compatibili con il valore atteso. Tuttavia le incertezze ottenute sono estremamente più piccole: per gli ultimi dati sono, approssimativamente, $2,6\times10^{-5}$. 
- Il rapporto tra le incertezze della prima simulazione e di quest'ultima vale circa $19$. Questo valore è in accordo con la formula generale della varianza precedentemente introdotta: $\sigma^2=\langle g^2 \rangle - \langle g \rangle^2$. $\langle g \rangle^2=1$ per entrambi gli esperimenti, mentre $\int_0^1\left(\frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right)\right)^2dx=\frac{\pi^2}{8}\simeq1,233$ e $\int_0^1\left(\frac{\pi}{2}\cos(\frac{\pi}{2}x)\frac{\pi^4-80\pi^2+1920}{1920\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right)}\right)^2\frac{1920}{\pi^4-80\pi^2+1920}\left(1-\frac{\pi^2x^2}{8}+\frac{\pi^4x^4}{384}\right) dx \simeq1,000672$. Il rapporto tra le varianze è circa $350$, da cui, estraendo la radice, si ricava che il rapporto atteso tra le varianze è $18,6$.
- Da queste semplici considerazioni si potrebbe assumere che il secondo metodo sia dunque estremamente migliore del primo, dal momento che si ottengono risultati venti volte più precisi. Tuttavia, bisogna tenere in considerazione anche il tempo computazionale. Utilizzando tecniche di rigetto, infatti, non può che aumentare: i numeri casuali effettivamente generati sono più di $M=10^6$. 
- Nel calcolo di integrali con tecniche Monte Carlo, si utilizza spesso un parametro che misura l'efficienza dell'algoritmo: $Q_i=t_i\sigma_i^2$. In questa formula $\sigma_i^2$ rappresenta la varianza, mentre $t_i$ il tempo computazionale. Un metodo è tanto più efficiente quanto più è piccolo $Q_i$. Per confrontare i due diversi metodi utilizzati, si dovrebbe stimare anche il rapporto tra i tempi computazionali. 
- Il suo ordine di grandezza si può calcolare, in modo molto grossolano, dal rapporto dei seguenti integrali:

- L'interpretazione geometrica del metodo del rigetto è la seguente: la coppia $(x;r)$ viene accettata se il punto $(x;p_{max}r)$ si trova nella parte di piano compresa tra l'asse delle ascisse e la $p(x)$. La probabilità che ciò accada è dunque data dal rapporto tra l'integrale di $p(x)$ e quello di $p_{max}$. Il numero medio di $x$ da generare, ogni volta, prima che possa essere accettata è dato dall'inverso di questo rapporto (è il valore che si può osservare nel disegno). Il tempo computazionale risulta quindi aumentato di circa un fattore $1,6$.
- Questa è soltanto una prima stima di $t_i$: volendo essere più precisi bisognerebbe tenere conto anche che, per applicare la tecnica del rigetto, serve generare, ogni volta, due numeri casuali (ovvero $x$ ed $r$). Bisognerebbe dunque moltiplicare per un ulteriore fattore due. Inoltre, anche per calcolare le varie $g(x_i)$ i tempi impiegati potrebbero non essere esattamente gli stessi. 
- In ogni caso, si può certamente affermare che questo secondo metodo sia il più efficiente: la varianza è stata ridotta di circa $350$ volte, mentre il tempo computazionale è dello stesso ordine di grandezza della prima simulazione. 
- Si ripete l'esperimento una terza volta, cercando però di utilizzare una $p(x)$ che possa essere invertita analiticamente. Eliminando le complicazioni del rigetto, è infatti possibile confrontare direttamente le efficienze dei metodi tramite il rapporto delle varianze. 
- Non si ha una vasta scelta per $p(x)$: se fosse un polinomio, dovrebbe essere necessariamente di primo grado, dal momento che non si sanno risolvere esattamente equazioni di grado superiore al secondo. Si decide quindi di sviluppare $\cos(\frac{\pi}{2}x)$ al primo ordine in $x=\frac{1}{2}$. Con la corretta normalizzazione risulta che:
$$p(x)=1-\frac{\pi}{2}\left(x-\frac{1}{2}\right) \quad \quad \bar{g}(x)=\frac{\pi}{2}\cos\left(\frac{\pi}{2}x\right)\frac{1}{1-\frac{\pi}{2}\left(x-\frac{1}{2}\right)}$$
- Invertendo la funzione cumulativa $F(x)=-\frac{\pi x^2}{4}+\frac{\pi x}{4}+x$ si ottiene che, se $y$ è uniformemente distribuito nell'intervallo $[0;1)$, $x=\frac{\pi+4-\sqrt{(\pi+4)^2-16\pi y}}{2\pi}$ è distribuito secondo la densità di probabilità richiesta.
- In questa fase dell'esperimento non vengono modificati i valori di $M$ e di $N$ e, ripetendo tutte le operazioni precedentemente descritte, si ottengono i seguenti risultati:

```python
i, intg, err = np.loadtxt("integrale3.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,intg,yerr=err)
plt.xlabel('numero passi')
plt.ylabel('I')
plt.grid(True)
plt.plot(x,y,color="red") 
plt.show()
```

- Anche in questo caso, i risultati sono sempre compatibili con il valore atteso dell'integrale entro $2\sigma$. Il valore ottenuto con tutti i cento blocchi è, tuttavia, leggermente più piccolo di uno, pur considerando la sua incertezza. Ciò non deve sorprendere eccessivamente: la distanza è estremamente piccola e può essere attribuita a fluttuazioni statistiche. Bisogna sempre ricordare, inoltre, che, se il risultato $x$ di una misura è distribuito secondo una curva gaussiana definita dal valore medio $\langle X \rangle$ e dalla deviazione standard $\sigma$, vi è una probabilità di circa $\frac{1}{3}$ che $x \notin [\langle X \rangle-\sigma;\langle X \rangle+\sigma]$ 
- In questo caso $\langle g^2 \rangle - \langle g \rangle^2\simeq0,01785$, pertanto le incertezze dovrebbero risultare $3,6$ volte di quelle ottenute tramite la distribuzione uniforme. Gli ultimi dati del grafico presentano un'incertezza di circa $1,3\times10^{-4}$. Effettuando il rapporto con le $\sigma$ del primo esperimento, si ottiene $3,8$. Anche in questo caso, quindi, il valore è molto simile a quanto atteso.
- Questi risultati, sono invece molto meno precisi di quelli ottenuti dal secondo esperimento. In questo caso, infatti, per poter invertire la cumulativa, ci si è dovuti accontentare di uno sviluppo al primo ordine della funzione. Sicuramente questa $p(x)$ "somiglia" di meno a $g(x)$ rispetto a quella precedente ricavata da uno sviluppo al quarto ordine.
- Si è dunque riusciti anche a verificare quantitativamente che, più la distribuzione di probabilità è simile all'integranda, minore sarà la varianza su $I$.

### **RANDOM WALK TRIDIMENSIONALE**
### Simulare un random walk e disegnare, in funzione del numero di step $i\in [0;100]$, $\sqrt{\langle |\vec{r}|^2 \rangle}$ con le relative incertezze nei due seguenti casi:
1. **RANDOM WALK SU UN RETICOLO CUBICO**, ad ogni step la particella può muoversi in avanti o indietro in **una** delle tre direzioni principali (si assume che ciascuna abbia la stessa probabilità)

2. **RANDOM WALK NEL CONTINUO**, ad ogni step la particella può muoversi in una generica direzione (si assume che ognuna abbia la stessa probabilità)
 


- Si assume che la particella parta dall'origine del sistema cartesiano e che, ad ogni step, si sposti di una quantità fissata $a=1$. Per poter ottenere risultati statisticamente significativi (ovvero corredati con le relative incertezze) si devono simulare parecchi random walk (in questo caso si è scelto $M=10^4$). 
- Fissando inizialmente il numero $i$ di step "percorsi", per poter analizzare i risultati dell'esperimento si utilizzano tre array (x,y,z) di dimensione $M$. Fissata una loro componente k, $(x[k];y[k];z[k])$ rappresenta la posizione nello spazio della particella durante la k-esima simulazione di random walk. Per ogni $M$ si può dunque calcolare $|\vec{r_j}|^2 $ ($j=1,2,\dots,M$) tramite il teorema di Pitagora.
- Per poter calcolare le incertezze si divide, ancora una volta, l'esperimento in $N=100$ blocchi. Ciascuno di essi contiene, quindi, $L=\frac{M}{N}$ simulazioni di random walk. Si ottengono $N$ stime del valore di $\langle |\vec{r_k}|^2 \rangle$ ($k=1,2,\dots,N$), calcolate dalla media degli $L$ $ |\vec{r_j}|^2 $ presenti in quel blocco. 
- Come negli esercizi precedenti il valore atteso di $\langle |\vec{r}|^2 \rangle$ è la media della delle cento stime e la sua incertezza si può calcolare mediante l'usuale formula.
- Se si vuole graficare $\sqrt{\langle |\vec{r}|^2 \rangle}$ è sufficiente estrarre la radice del valore atteso e, per calcolare la $\sigma$, utilizzare la teoria di propagazione degli errori. Se una grandezza è funzione di un'altra (di cui è nota la deviazione standard) $y=y(x)$, vale che $\sigma_y=\left\lvert{\frac{dy(x)}{dx}}\right\lvert\sigma_x$. Per cui in questo caso risulta che $\sigma_r=\frac{1}{2\sqrt{\langle |\vec{r}|^2 \rangle}}\sigma_{r^2}$
- Si ripete la stessa analisi per tutti gli step $i=1,2,\dots,100$ che si desidera analizzare. Inizialmente (dopo zero step) tutte le componenti degli array x,y,z sono pari a zero (tutte le particelle partono dall'origine). Serve, a questo punto, un algoritmo che permetta di aggiornare le componenti dei tre array a seguito di uno step. 
- Per il random walk sul reticolo si genera un numero casuale (da una distribuzione uniforme) $v\in[0;6)$ che determina la direzione (e il verso) dello spostamento. Per convenzione, si è deciso che, se la parte intera di v è pari a zero, lo spostamento avverrà di $a$ nel verso positivo dell'asse x; se $\lfloor{v}\rfloor=1$, allora lo spostamento sarà nel verso negativo di x. Si procede (ordinatamente) con la stessa logica per gli altri quattro valori possibili di $\lfloor{v}\rfloor$. 
- Questa scelta è, ovviamente, del tutto arbitraria: si potrebbe associare in modo diverso la direzione dello spostamento alla parte intera del numero casuale (ad esempio partendo dall'asse z). Tuttavia, poiché si utilizza una distribuzione uniforme, i risultati finali non cambiano, specialmente quando l'esperimento viene ripetuto diverse volte. Il procedimento viene ripetuto per ciascuno degli $M$ random walk che si desidera simulare. 
- Per il random walk continuo, invece, è sufficiente determinare la densità di probabilità che campioni uniformemente l'angolo solido. Si può dimostrare che questa distribuzione si ottiene generando $\phi\in[0;2\pi)$ uniformemente e $\theta=\arccos{(1-2r)}$  ($r\in[0;1)$ unifomemente). A questo punto, con la coppia $(\theta;\phi)$, tramite le coordinate sferiche, si possono aggiornare le tre componenti di ciascuna posizione. Anche in questo caso si ripete il procedimento $10^4$ volte.
-
