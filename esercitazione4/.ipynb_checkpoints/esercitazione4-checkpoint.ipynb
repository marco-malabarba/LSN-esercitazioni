## <span style="color:blue">ESERCIZIO 4.1 </span>

Nel file di input è stato inserita la variabile "restart" che, se uguale ad uno, permette al programma di partire da una configurazione precedente (anziché da quella con le particelle disposte sul reticolo cristallino).

Il codice fornito è stato leggermente modificato: si è stampato sul file "old.0" la posizione degli atomi al penultimo step dell'algoritmo. In questo modo, nel momento in cui si desidera far ripartire la simulazione, si hanno a disposizione le $\vec{r}(t-dt)$ e le $\vec{r}(t)$ (salvate su "old.final"). 

Viene poi letta la temperatura $T^*$ (sempre in "input.dat"). Se $T^*=0$, il programma comincia semplicemente ad utilizzare l'algoritmo di Verlet per studiare l'evoluzione del sistema. Ciò equivale, di fatto, ad allungare la simulazione precedendemente terminata.

Se, invece, $T*$ è diversa da zero, si applica l'algoritmo di Verlet una sola volta per determinare $\vec{r}(t+dt)$. Sfruttando le funzioni fornite, si calcolano le velocità delle particelle e, di conseguenza, l'energia cinetica e la temperatura del sistema $T_{sist}^*$. Si valuta il rapporto $p=\frac{T^*}{T^*_{sist}}$ e si scalano le velocità per il fattore $\sqrt{p}$. Si ricalcolano, infine, le posizioni $\vec{r}(t)$ tramite l'equazione $\vec{r_{new}}(t)=Pbc(\vec{r}(t+dt)-vdt)$. Nella formula precedente le velocità sono quelle già riscalate di $\sqrt{p}$. Si sono utilizzate le periodic boundary conditions per evitare di spostare, irrimediabilmente, una o più particelle fuori dal box di interesse.

In questo modo si forza il sistema ad avere, almeno in principio, la $T^*$ desiderata. Ciò è stato ampiamente sfruttato nelle fasi di equilibrazione dei tre stati di aggregazione. Partendo, infatti, dalla configurazione fcc, si può notare che la temperatura subisce un brusco crollo nei primissimi step dell'algoritmo. Disordinando il reticolo perfetto, l'energia potenziale aumenta (diventa meno negativa) e, per la conservazione dell'energia, l'energia cinetica e la temperatura calano di conseguenza.

## <span style="color:blue">ESERCIZIO 4.2 </span>

Per introdurre le incertezze sulle quattro grandezze, si utilizza la media a blocchi. Per tutti gli stati di aggregazione si è deciso di porre $n_{blocchi}=100$ (il valore di $n_{step}$ verrà invece discusso più avanti). In analogia con le precedenti esercitazioni, si sono utilizzati due array: uno contiene le stime dei $100$ blocchi, l'altro i valori del primo elevati al quadrato. In questo caso gli array sono bidimensionali: la seconda componente identifica una delle quattro variabili, mentre la prima si riferisce, normalmente, ad uno dei cento blocchi.

In questo modo, su altri vettori, si possono calcolare le medie cumulative e le relative incertezze delle quantità termodinamiche, in funzione del numero di blocchi considerati. 

Nel file di input è stata inserita la variabile "analisi". Se è pari a zero, tutti i calcoli precedentemente descritti non vengono effettuati dal programma. Per le fasi iniziali di equilibrazione, infatti, non si è interessati a uno studio approfondito delle varie quantità termodinamiche, ma si vogliono solo effettuare simulazioni rapide.   

## <span style="color:blue">ESERCIZIO 4.3 </span>

A questo punto, dopo tutte i passaggi preliminari descritti nel primo esercizio, si sono studiati i tre stati di aggregazione. Per lo stato solido e quello liquido si è utilizzato un numero di passi totali $n_{step}=10^5$. È stato facile notare, invece, che questo valore non era adeguato per il gas. I risultati che si ottenevano erano, infatti, piuttosto insoddisfacenti: in alcune simulazioni le medie cumulative avevano un andamento monotono in regioni estremamente ampie e le incertezze, anziché diminuire, crescevano all'aumentare del numero dei blocchi considerati. Il problema si è risolto ponendo $n_{step}=10^6$. Probabilmente, un numero di step pari a $10^5$ non è sufficiente per valutare correttamente il comportamento del sistema gassoso e, quindi, per stimare gli andamenti delle quantità termodinamiche di interesse.

Finora si sono utilizzate le unità di misura adimensionali di Lennard-Jones (identificate con un asterisco in apice alla grandezza). Con questa scelta si sono evitati problemi di underflow nel programma. Per convertire una $T^*$ (e la sua incertezza) in unità del sistema internazionale, è sufficiente moltiplicarla per $\frac{\epsilon}{k_B}$. La conversione delle energie si effettua, invece, moltiplicando per $\epsilon$. Per gli atomi di argon $\frac{\epsilon}{k_B}=120K$.

Nei grafici sottostanti verranno mostrati, in unità SI, all'aumentare del numero di misure $n_{mis}$, i valori medi di: temperatura; energia cinetica per particella; energia potenziale per particella ed energia totale per particella. Considerando un numero di blocchi pari, ad esempio, a $k$, si dimostra facilmente che $n_{mis}=\frac{n_{step}k}{10n_{blocchi}}$. Il fattore $10$ è stato inserito poiché le misure vengono effettuate ogni dieci step dell'algoritmo di Verlet.      

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def grafici(stato):
	if stato=="solido":
		titolo1="solido"
	if stato=="liquido":
		titolo1="liq"
	if stato=="gassoso":
		titolo1="gas"
		
	titolo2="temp"
	titolo="average_"+titolo2+"_"+titolo1+".dat"
	
	step, t,err = np.loadtxt(titolo, usecols=(0,1,2), delimiter=' ', unpack='true')
	t=t*120.
	err=err*120.
	plt.errorbar(step,t,yerr=err)
	plt.title ('Temperatura')
	plt.xlabel('numero di misure')
	plt.ylabel('T[K]')
	plt.grid(True)
	plt.show()
	
	titolo2="ekin"
	titolo="average_"+titolo2+"_"+titolo1+".dat"
		
	step, ekin,err = np.loadtxt(titolo, usecols=(0,1,2), delimiter=' ', unpack='true')
	ekin=ekin*120.*1.380649*(10**(-23))
	err=err*120.*1.380649*(10**(-23))
	plt.errorbar(step,ekin,yerr=err)
	plt.title ('Energia cinetica per particella')
	plt.xlabel('numero di misure')
	plt.ylabel('$<K/N> [J]$')
	plt.grid(True)
	plt.show()
	
	titolo2="epot"
	titolo="average_"+titolo2+"_"+titolo1+".dat"
	
	step, epot,err = np.loadtxt(titolo, usecols=(0,1,2), delimiter=' ', unpack='true')
	epot=epot*120.*1.380649*(10**(-23))
	err=err*120.*1.380649*(10**(-23))
	plt.errorbar(step,epot,yerr=err)
	plt.title ('Energia potenziale per particella')
	plt.xlabel('numero di misure')
	plt.ylabel('<U/N> [J]')
	plt.grid(True)
	plt.show()
	
	titolo2="etot"
	titolo="average_"+titolo2+"_"+titolo1+".dat"
	
	step, etot,err = np.loadtxt(titolo, usecols=(0,1,2), delimiter=' ', unpack='true')
	etot=etot*120.*1.380649*(10**(-23))
	err=err*120.*1.380649*(10**(-23))
	plt.errorbar(step,etot,yerr=err)
	plt.title ('Energia totale per particella')
	plt.xlabel('numero di misure')
	plt.ylabel('<E/N> [J]')
	plt.ticklabel_format(useOffset=False)
	plt.grid(True)
	plt.show()
```

```python
print("Stato solido")
grafici("solido")
```

```python
print("Stato liquido")
grafici("liquido")
```

```python
print("Stato gassoso")
grafici("gassoso")
```

Tutti i risultati sembrano piuttosto verosimili. Gli andamenti dell'energia cinetica sono analoghi a quelli della temperatura. Tra le due quantità vi è infatti una semplice relazione di proporzionalità: dalla termodinamica è noto che, per una particella, $\left\langle K\right\rangle=\frac{3}{2}k_BT$. 

Osservando le scale dei grafici, si può concludere che l'energia totale è la grandezza che fluttua di meno. L'algoritmo di Verlet è, infatti, rinomato per la sua abilità nel mantenere approssimativamente costante la $E$, specie su scale di tempo lunghe.

Lo stato solido è caratterizzato da un'energia potenziale molto negativa e da un'energia cinetica piccola: gli atomi di argon sono disposti in posizioni specifiche dello spazio e le loro oscillazioni sono piuttosto contenute. Di conseguenza, la $\left\langle E/N\right\rangle$ è piuttosto negativa. Al contrario, il gas possiede un'alta $\left\langle K/N\right\rangle$ e un'energia potenziale prossima a zero: le particelle si muovono quasi liberamente nello spazio e, quindi, $\left\langle E/N\right\rangle$ risulta positiva. Lo stato liquido è caratterizzato da proprietà intermedie tra il solido e il gas.      

Come previsto, non in tutte e tre le fasi si ottengono delle $T$ medie compatibili, entro le incertezze, con la $T_{target}$. Per lo stato liquido, infatti, la temperatura $T_{liq}$ è leggermente superiore a quella desiderata. Il risultato è, comunque, del tutto verosimile: la discrepanza è di circa $2\sigma$, l'errore percentuale è ben inferiore all'$1\%$, inoltre la $T_{liq}$ è decisamente distante da quelle caratteristiche degli altri due stati.
