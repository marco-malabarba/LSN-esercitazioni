## <span style="color:blue">ESERCIZIO 12.1</span>

Inizialmente si è provato ad utilizzare il codice che era stato fornito senza alcuna sostanziale modifica. Ci si può rendere conto, però, che il numero di epoche può essere aumentato fino a $N_{epochs}=25$ senza incorrere in overfitting. L'andamento della loss sui dati di validation rimane infatti decrescente all'aumentare delle epoche.

Questi sono stati i risultati ottenuti:

```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
seed=0
np.random.seed(seed) 
tf.random.set_seed(seed)

from tensorflow.keras.datasets import mnist
img_rows, img_cols = 28, 28 

num_classes = 10 

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train /= 255
X_test /= 255


Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

def create_DNN():
    model = Sequential()
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

print('Model architecture created successfully!')

from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

def compile_model(opt):
    model=create_DNN()
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=opt,
                  metrics=['accuracy'])
    return model

print('Model compiled successfully and ready to be trained.')
```

```python
batch_size = 32
epochs = 25

opt='SGD'
model_DNN = compile_model(opt)

history = model_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
score = model_DNN.evaluate(X_test, Y_test, verbose=1)

print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])


plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

Ci si può ritenere soddisfatti dei risultati ottenuti: si è riusciti ad ottenere, al termine delle 25 epoche, un valore della loss piuttosto basso e un'accuratezza quasi del $98\%$.

Si vuole provare a testare questa intelligenza artificiale con delle cifre scritte tramite l'applicazione gimp. Molto probabilmente alcune predizioni saranno errate: la macchina è stata allenata con dei numeri scritti con una calligrafia diversa.

```python
from PIL import Image
import os

for ii in range(10):
    string1=str(ii)
    digit_filename = string1 + ".png"
    digit_in = Image.open(digit_filename).convert('L')

    ydim, xdim = digit_in.size
    pix=digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for i in range(xdim):
            data[i,j]=pix[j,i]

    data /= 255

   
    data = data.reshape(1,xdim*ydim)
    pred_0 = model_DNN.predict(data)
    data = data.reshape(xdim,ydim)
    if np.argmax(pred_0)==ii: #argmax perché l'output è un array
        print(ii,end="")
        print(": corretto")
    else:
        print(ii,end="")
        print(": errato (",end="")
        print(np.argmax(pred_0),end="")
        print(")")
```

Come previsto solo sette delle dieci cifre sono state interpretate correttamente. Nel programma precedente i numeri tra parentesi sono le previsioni, errate, dell'intelligenza artificiale.

Si prova ora a cambiare l'ottimizzatore: anziché il "classico" stocastic gradient descent, si decide di utilizzare NADAM. Anche in questo caso, dopo aver terminato la fase di training, vengono sottoposte alla macchina le cifre scritte con gimp.

```python
batch_size=32
epochs=7
opt="Nadam"
model_DNN=compile_model(opt)
history = model_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
print()
print("OTTIMIZZATORE ",end="")
print(opt)
print()
score = model_DNN.evaluate(X_test, Y_test, verbose=1)

print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])


plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
for ii in range(10):
    string1=str(ii)
    digit_filename = string1 + ".png"
    digit_in = Image.open(digit_filename).convert('L')

    ydim, xdim = digit_in.size
    pix=digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for i in range(xdim):
            data[i,j]=pix[j,i]

    data /= 255

   
    data = data.reshape(1,xdim*ydim)
    pred_0 = model_DNN.predict(data)
    data = data.reshape(xdim,ydim)
    if np.argmax(pred_0)==ii: #argmax perché l'output è un array
        print(ii,end="")
        print(": corretto")
    else:
        print(ii,end="")
        print(": errato (",end="")
        print(np.argmax(pred_0),end="")
        print(")")
```

Si possono notare notevoli differenze: dopo appena sette epoche ci si è dovuti fermare poiché iniziava ad insinuarsi l'overfitting. Inoltre, dopo appena un'epoca, si erano già raggiunti degli ottimi risultati (l'accuratezza sfiorava il $97\%$).

Tuttavia, al termine del training, i risultati sono piuttosto simili a quelli ottenuti con lo SGD: solo sette delle dieci cifre scritte a mano sono state predette correttamente e i valori finali di loss e accuracy sono comparabili con i precedenti.

Si modifica ancora l'ottimizzatore, questa volta viene adoperato ADAMAX.

Per non andare in overfitting si è posto $N_{epochs}=10$. 

Si può osservare che questi risultati sono leggermente migliori dei precedenti: l'accuratezza finale supera il $98\%$ e sono state interpretate correttamente otto cifre su dieci. Inoltre, a differenza di NADAM, si osserva più chiaramente il progresso nelle varie epoche. Dopo la prima, infatti, la loss era ancora relativamente alta (sia sui dati di training che su quelli di validation).

Sulla base di queste considerazioni, si potrebbe erroneamente immaginare che lo SGD sia l'ottimizzatore più "elementare" e, scegliendone un qualsiasi altro, i risultati che si ottengono sono equivalenti o migliori. Quest'affermazione è facilmente confutabile se si decide si adoperare ADADELTA.

Si evince che la macchina performa in modo notevolmente peggiore. L'accuratezza finale rimane, infatti, ben al di sotto del $90\%$ e solo sei cifre sono state correttamente interpretate. È interessante notare, inoltre, che la convergenza di ADADELTA è molto più lenta: al termine della prima epoca l'accuracy sui dati di training era appena del $15\%$, non molto diversa da quella che si conseguirebbe assegnando ai numeri una label casuale.  

Anche in questa esercitazione ci si rende quindi conto che la scelta dell'ottimizzatore non è affatto banale: non tutti sono adeguati per uno specifico problema e ciascuno ha la propria velocità di convergenza.

## <span style="color:blue">ESERCIZIO 12.2</span>

Partendo da una rete simile a quella reperibile sul sito web https://www.cs.ryerson.ca/~aharley/vis/conv/flat.html, si sono modificati varie volte i parametri al fine di massimizzare l'accuratezza della macchina nelle previsioni.

La rete convoluzionale creata è così strutturata: nel primo layer si applica la convoluzione bidimensionale con sei filtri $5x5$. Si è poi eseguito un max-pooling di dimensione $2x2$, seguito da un'altra convoluzione con sedici filtri $5x5$. Dopo un ulteriore max-pooling $2x2$, si è "srotolata" la rete e si è creato un altro layer, prima di quello finale di output, di 256 neuroni. 

Per i dieci neuroni del layer di output si è sfruttata la funzione di attivazione SOFTMAX, mentre, per tutti gli altri (esclusi ovviamente quelli di max-pooling), la RELU.

Come per la deep neural network, la loss è la categorical cross entropy. Si è invece adoperato ADAMAX come ottimizzatore, dal momento che aveva fornito i risultati più soddisfacenti con la deep neural network.

I dati di training e di validation utilizzati sono, anche in questo caso, quelli forniti da MNIST.

Si è posto $N_{epochs}=10$: aumentando ulteriormente questo valore si incappa nell'overfitting. 

```python
from tensorflow.keras.layers import Flatten, Conv2D, MaxPooling2D

if tf.keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
    
def create_CNN(opt):
    model = Sequential()
    model.add(Conv2D(6, kernel_size=(5, 5),activation='relu',input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2,2),))
    model.add(Conv2D(16, kernel_size=(5,5),activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Flatten())
    model.add(Dense(256,activation='relu'))
    model.add(Dense(10,activation='softmax'))
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=opt,
                  metrics=['accuracy'])
    return model
```

```python
batch_size=32
epochs=10
opt="Adamax"
model_CNN=create_CNN(opt)
history = model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
print()
print("OTTIMIZZATORE ",end="")
print(opt)
print()
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])


plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
for ii in range(10):
    string1=str(ii)
    digit_filename = string1 + ".png"
    digit_in = Image.open(digit_filename).convert('L')

    ydim, xdim = digit_in.size
    pix=digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for i in range(xdim):
            data[i,j]=pix[j,i]

    data /= 255

   
    data = data.reshape(1,xdim*ydim)
    pred_0 = model_DNN.predict(data)
    data = data.reshape(xdim,ydim)
    if np.argmax(pred_0)==ii: #argmax perché l'output è un array
        print(ii,end="")
        print(": corretto")
    else:
        print(ii,end="")
        print(": errato (",end="")
        print(np.argmax(pred_0),end="")
        print(")")
```

I risultati ottenuti sono dolceamari. Da un lato si può notare che l'accuratezza finale sui dati di validation ottenuta arriva a sfiorare il $99\%$, molto meglio di tutte le prove effettuate precedentemente con la deep neural network. D'altro canto, però, solo sei cifre su dieci sono state interpretate correttamente. 

Si è tentato, allora, di modificare l'ottimizzatore e di cambiare leggermente la struttura della rete stessa, ma, curiosamente, gli errori permanevano sempre sulle stesse cifre (a fronte di un'accuratezza inferiore). Come già accennato precedentemente, l'intelligenza artificiale è stata allenata con numeri scritti con una calligrafia diversa, non deve quindi sorprendere che possano esserci degli errori sulle cifre disegnate con gimp.
