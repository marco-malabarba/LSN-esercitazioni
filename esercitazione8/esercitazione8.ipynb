Gli esercizi sono stati risolti nel programma "variazionale.cxx". 

Anche per questo problema si utilizza l'algoritmo di Metropolis. Data una funzione d'onda di prova $\Psi_{T}$ si può dimostrare che: $$\langle {\hat H} \rangle_T = \int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}$$

Questo integrale si presenta quindi nella forma "standard" $\int dx p(x)f(x)$ con $p(x)$ generica distribuzione di probabilità. Tuttavia, sarebbe quasi sempre impossibile da calcolare con metodi Monte Carlo elementari poiché la $p(x)$ è illimitata. Con il Metropolis, invece, l'esercizio si può risolvere piuttosto agilmente. Innanzitutto si può notare che l'integrale del modulo quadro della $\Psi_T$ è trascurabile, in quanto è una semplice costante che si elide nel rapporto tra la $|\Psi_T|^2$ nel nuovo e nel vecchio punto. Ciò non è fondamentale in questo caso: $\int dx |\Psi_T(x)|^2$ è calcolabile analiticamente se si utilizza, come funzione d'onda di prova, la somma di due gaussiane simmetriche rispetto all'asse y. L'osservazione sarebbe stata cruciale, invece, se si fosse scelta una $\Psi_T$ più complicata avente una normalizzazione non nota. Campionando correttamente la distribuzione di probabilità $|\Psi|^2$, è poi immediato calcolare $\langle {\hat H} \rangle_T$. Dalla meccanica quantistica è noto che: $$\hat H \Psi_T(x)=-\frac{\hbar^2}{2m}\frac{\partial^2\Psi_T}{\partial x^2}+V(x)\Psi_T$$

Per evitare problemi di underflow si è posto $\hbar=m=1$. Nell'espressione precedente la derivata seconda si può sempre calcolare analiticamente, indipendentemtente dalla difficoltà della $\Psi_T$ (a differenza del suo integrale). Con $\Psi_T^{\sigma,\mu}(x) = e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}$, si ha che $$\frac{\partial^2\Psi_T}{\partial x^2}=\dfrac{\frac{\left(x+{\mu}\right)^2\mathrm{e}^{-\frac{\left(x+{\mu}\right)^2}{4{\sigma}^2}}+\left(x-{\mu}\right)^2\mathrm{e}^{-\frac{\left(x-{\mu}\right)^2}{4{\sigma}^2}}}{2{\sigma}^2}-\mathrm{e}^{-\frac{\left(x+{\mu}\right)^2}{4{\sigma}^2}}-\mathrm{e}^{-\frac{\left(x-{\mu}\right)^2}{4{\sigma}^2}}}{2{\sigma}^2}$$.

Ad ogni step del Metropolis la $x_{new}$ veniva estratta casualmente (con una distribuzione uniforme) tra $[x_{old}-2;x_{old}+2]$. Con questa larghezza si è osservato, infatti, che l'accettazione era prossima al $50\%$ per dei range piuttosto ampi sia di $\mu$ che di $\sigma$. Inoltre, questo valore è ragionevole in quanto compatibile con la scala su cui le $\Psi_T$ si mantengono significativamente diverse da zero.

Si è effettuato uno studio preliminare ponendo $\mu=0$ e $\sigma=1$. Si è scelto di partire da $x=0$. Si sono poi effettuati $10000$ step di equilibrazione per il sistema. Questo numero poteva essere notevolmente ridotto o addirittura azzerato: $x=0$ è, in realtà, il massimo della distribuzione $|\Psi_{\mu=0; \sigma=1}|^2$. Tuttavia, questa fase non richiede grosso sforzo computazionale ed è completata in un tempo molto piccolo. Inoltre, se si decidesse di cambiare i valori di $\mu$ e $\sigma$, si avrebbe così la certezza di campionare correttamente la $|\Psi|^2$.  Con $100000$ step si è ottenuta una prima stima (per ora priva di incertezze) di $\langle {\hat H} \rangle_{\mu=0; \sigma=1}\simeq-0,24$. 

È noto che lo stato fondamentale di questo sistema quantistico si trova ad un'energia $E_{gs}\simeq-0,46$. Bisogna determinare per quali valori di $\mu$ e di $\sigma$ si ottiene la migliore approssimazione del ground state. Si decide, allora, di calcolare la $\langle {\hat H} \rangle$, sempre con $M=100000$ step, al variare dei parametri $\mu$ e $\sigma$. Entrambi vengono variati nell'intervallo $[0,1;1]$ con step di $0,1$. Si ottengono dunque $100$ stime del valore medio dell'energia. Si seleziona la coppia $(\mu;\sigma)$ che minimizza $\langle {\hat H} \rangle$. Poiché, al termine di questa prima fase, si sono ottenuti $\mu_{best}=0,8$ e $\sigma_{best}=0,6$, la ricerca è poi proseguita sulla successiva cifra decimale (ovvero si sono variati $\mu\in[0,71;0,89]$ e $\sigma\in[0,51;0,69]$ con step di $0,01$). Ripetendo questo stesso algoritmo qualche altra volta, si sono ottenuti, in definitiva, come migliori parametri $\mu_{best}=0,802393$ e $\sigma_{best}=0,6184$. 

Si è deciso di non proseguire oltre con le cifre decimali in quanto questo metodo non è esatto. Tutti i valori degli integrali sono affetti, infatti, da un'incertezza. È logico attendersi che, all'aumentare delle cifre significative dei parametri, le $\langle {\hat H} \rangle$ diventino sempre più simili tra loro. Poiché si utilizza un approccio numerico, c'è la possibilità che la coppia $(\mu;\sigma)$, che dovrebbe minimizzare la $\langle {\hat H} \rangle$, risulti avere, invece, un'energia media più elevata delle configurazioni "contigue" (a causa di fluttazioni statistiche). Si sarebbero potute calcolare le incertezze associate a tutti questi integrali, ma avrebbero solamente aumentato il tempo computazionale e sarebbero state di scarso aiuto. Per determinare correttamente in che direzione muoversi nella ricerca dei parametri, servirebbe conoscere il valore vero di $\langle {\hat H} \rangle$. Avendo a disposizione solamente un intervallo, in cui verosimilmente si trova il valore dell'integrale, invece, l'approccio più logico da seguire è quello di considerarne il punto medio. Questo corrisponde, in realtà, alla stima (senza incertezze) di $\langle {\hat H} \rangle$.

Si è ripetuto il calcolo dell'integrale con i parametri migliori. In questo caso si è però aumentato il numero degli step totali a $M=10^8$. Per ottenere una stima $\langle {\hat H} \rangle$ corredata da un'incertezza, si è divisa, come nelle precedenti esercitazioni, la simulazione in $N=100$ blocchi. 

Nel grafico sottostante viene mostrato, come al solito, l'andamento della $\langle {\hat H} \rangle$ (e del relativo errore) all'aumentare del numero dei blocchi considerati. Si disegna, inoltre, un istogramma contenente la distribuzione $|\Psi(x)^2|$ campionata.  
  

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def f(x,mu,sigma): 
	psi=np.exp(-(x-mu)**2/(2*(sigma**2))) + np.exp(-(x+mu)**2/(2*(sigma**2))) #funzione d'onda non normalizzata
	norm=2*np.sqrt(np.pi)*sigma*np.exp(-mu*mu/(sigma*sigma))+2*np.sqrt(np.pi)*sigma; #fattore di normalizzazione
	return (psi**2)/norm


i, h, err = np.loadtxt("ground_state.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,h,yerr=err)
plt.title("Andamento <H> con i migliori parametri")
plt.xlabel('numero blocchi')
plt.ylabel('$<H>$')
plt.grid(True)
plt.show()

mu=0.802393
sigma=0.618400
N = 1000
x = np.linspace(-3,3,N,endpoint=True)
fx=f(x,mu,sigma)
x_med,occ=np.loadtxt("histo.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.hist(x_med, weights=occ,bins=len(x_med), density=1,color="gray") #density=1 --> normalizzazione automatica
plt.plot(x,fx,label="metodo variazionale",color="blue")

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
plt.plot(x,(psi[0])**2,color="red",label="metodo matriciale")
legend = plt.legend(loc='lower right',shadow=True)

plt.title("Istogramma della distribuzione $|\Psi(x)|^2$")
plt.xlabel('x')
plt.ylabel('$|\Psi(x)|^2$')
plt.grid(True)
plt.show()
```

Dal primo grafico si può osservare che il valore medio dell'Hamiltoniana sugli stati di prova $|\Psi_T>$ è superiore rispetto al ground state del sistema quantistico (che dovrebbe essere pari a circa $-0,46$). Ciò è conforme con quanto atteso: si può dimostrare che, dato un qualunque stato, risulta sempre $E_{gs}\le\langle {\hat H} \rangle$. 

Si può essere moderatamente soddisfatti del risultato ottenuto. Trascurando le incertezze, l'errore percentuale è del $3,5\%$. Ovviamente, in termini di unità $\sigma$ la compatibilità tra $E_{gs}$ e $\langle {\hat H} \rangle$ è pessima. La discrepanza è determinata dalla funzione d'onda di prova scelta. Infatti, nonostante si siano cercati i parametri migliori, la $\Psi_{gs}(x)$ non è perfettamente approssimabile tramite una somma di gaussiane. 

Si può dimostrare che, se il potenziale $V(x)\sim x^4$ per $x\to\infty$, allora $\Psi_{gs}(x)\sim \exp({-|x|^3})$. Di conseguenza, nel programma "variazionale_2.cxx" si è provato ad utilizzare come funzione d'onda di prova $\Psi_T^{\alpha,\mu}(x) = e^{-\alpha|(x-\mu)|^3}+e^{-\alpha|(x+\mu)|^3}$ (con $\alpha>0$). Non si è raggiunta, tuttavia, una migliore stima di $E_{gs}$. Nonostante il comportamento asintotico, infatti, nelle regioni vicine all'asse delle ordinate la somma delle gaussiane rimane la scelta migliore.

Nel secondo grafico si può confrontare la $|Psi_T(x)|^2$ ottenuta tramite il metodo variazionale con quella ricavata con il metodo matriciale (che è più preciso). L'istogramma (in grigio) è stato disegnato dividendo l'intervallo $[-3;3]$ in $60$ bin. Al di fuori di questo range, infatti, la distribuzione è estremamente prossima a zero. Per riempire i vari bin si sono utilizzati $10^7$ step dell'algoritmo di Metropolis. La normalizzazione è stata poi eseguita automaticamente dal programma in python con il comando "density=1". In blu, invece, è stato graficato direttamente il modulo quadro della funzione di prova con i parametri ottimizzati. È immediato rendersi conto che coincide perfettamente con l'istogramma. Ciò fornisce un'ulteriore evidenza della grandissima potenza dell'algoritmo di Metropolis: con poche righe di codice si riesce a campionare perfettamente una qualunque distribuzione di probabilità. In rosso è invece mostrato il modulo quadro della funzione d'onda ricavata con il metodo matriciale. Si può osservare che non coincide perfettamente con l'istogramma, nonostante gli andamenti generali siano piuttosto simili. Come già spiegato, è a causa di queste discrepanze che, con la somma di due gaussiane simmetriche, non si riesce ad ottenere un valore di $\langle {\hat H} \rangle$ più prossimo all'energia di ground state.
