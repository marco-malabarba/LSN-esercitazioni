## IL PROBLEMA DEL COMMESSO VIAGGIATORE

Per risolvere il problema si è implementata una classe intitolata "popolazione". Tra i data membri privati vi sono, innanzitutto, due array contenenti le posizioni, generate casualmente, delle 32 città (sulla circonferenza o nel quadrato). In questo modo, a ciascuna di essa è assegnato univocamente un indice $i=0,\dots,31$.

Si sono utilizzati 100 individui nella realizzazione dell'algoritmo genetico. Le informazioni fondamentali dell'intera popolazione sono state inserite in "m_individui", un array bidimensionale di interi (anch'esso tra i data membri provati). La prima componente permette di identificare l'individuo al quale ci si riferisce, mentre la seconda lo step del viaggio. Ad esempio l'istruzione m_individui[j][k]=i significa che il commesso j-esimo, dopo aver percorso k passi del suo percorso, si trova nella città i-esima (di cui si può ricavare la posizione tramite gli altri due array). Ovviamente $0\le j<100$ e $0\le k<32$.

Per ciascuno dei 100 individui si è poi calcolata la corrispettiva $L^{(2)}$ sommando tutte le distanze (elevate al quadrato) percorse nel suo viaggio. Si sono tenute in considerazione le condizioni al contorno del problema: il viaggiatore, al termine dell'itinerario, ritorna nella sua città natale.

Si è implementata, inoltre, una funzione che permette alla popolazione di riprodursi. In analogia con le teorie evolutive biologiche, gli individui che si adattano meglio al mondo esterno hanno maggiore probabilità di sopravvivere, e, quindi, di lasciare discendenti. In questa particolare applicazione, i commessi "migliori" sono quelli con la $L^{(2)}$ minore. Per assegnare le probabilità di riproduzione, si è calcolata, per ogni individuo, $k_i=\frac{1}{(L^{(2)})^2}$ e la si è poi divisa per la somma di tutti e cento i $k_i$ in modo da avere una corretta normalizzazione. Si è fatta questa scelta a seguito di numerosi tentativi. Utilizzando come $k_i$ il semplice reciproco di $L^{(2)}$, non venivano sufficientemente favoriti gli individui con la lunghezza minore: la convergenza era estremamente più lenta e nemmeno dopo migliaia e migliaia di generazioni si osservavano risultati accettabili. D'altro canto, aumentando ulteriormente l'esponente al denominatore, ad esempio ponendo $k_i=\frac{1}{(L^{(2)})^3}$, si verificava l'esatto opposto: quasi tutti i figli erano la copia esatta dell'individuo migliore della popolazione precedente. Ciò limita molto l'evoluzione del sistema, specialmente nelle primissime generazioni. Per ottenere commessi con un percorso più breve, bisogna, a quel punto, affidarsi alle mutazioni favorevoli (discusse più avanti), che sono però molto rare.

Vengono poi estratti due numeri casuali tra zero ed uno. Ciascuno di essi appartiene al range di probabilità di un individuo che viene dunque selezionato come genitore. Per semplicità due genitori danno sempre alla luce due figli. Non si è esclusa la possibilità che un individuo possa riprodursi con se stesso; ciò accade più facilmente se un commesso ha un percorso assai più breve di tutti gli altri e, di conseguenza, una probabilità di riproduzione molto grande. Da un punto di vista biologico, quindi, gli organismi possono fare sia riproduzione sessuata che asessuata.

In ogni caso, dati due genitori, il primo figlio segue per le prime sedici città il percorso del primo genitore, mentre nella seconda parte del viaggio vengono aggiunte le località rimanenti nell'ordine in cui si presentano nel secondo genitore. Per il secondo figlio, invece, la prima parte dell'itinerario è determinata dal secondo genitore.

Questo processo viene ripetuto 50 volte in modo tale che il numero di individui nella popolazione resti costante al variare delle generazioni. È importante ribadire che non tutti gli individui si riproducono: quelli con piccola $L^{(2)}$ verranno, più verosimilmente, selezionati varie volte nella scelta dei genitori.

Infine, si sono implementate le quattro mutazioni suggerite: permutazione di una coppia di città, shift in avanti di n città, permutazioni tra n città contigue, inversione dell'ordine di n città. In analogia con la biologia, non tutte le mutazioni accadono con la stessa probabilità, quelle più radicali (come la terza e la quarta) sono notevolmente più rare. Si è dunque posta pari a $0,1$ la probabilità della permutazione di coppia; per lo shift $p=0,05$; per la permutazione di più città e scelta per l'inversione $p=0,03$. Nelle ultime tre mutazioni, il valore di n è, anch'esso, determinato in modo casuale (rispettando ovviamente i limiti intuitivi: $1<n<\frac{n_{cities}}{2}$).

**NOTA:** l'implementazione della riproduzione e delle mutazioni può apparire banale e immediata (in quanto si tratta, in fin dei conti, di un esercizio sugli array), ma si è rivelata piuttosto ostica. Non è stato facile, in particolare, rispettare la condizione del problema che impone che ogni città debba essere visitata una sola volta. Per questo motivo, si è verificata preliminarmente la correttezza di queste funzioni utilizzando array di piccole dimensioni, in modo da controllare più facilmente le operazioni eseguite dal programma.


Per simulare l'evoluzione si sono, innanzitutto, creati 100 individui casuali. Il primo di essi percorre le città nell'ordine in cui sono state generate. Per gli altri 99, invece, si è scambiata la posizione di due città casuali dall'itinerario del commesso precedente.

A questo punto sono state calcolate le probabilità di riproduzione; si sono generati, seguendo l'algoritmo precedentemente descritto, i cento figli che hanno rimpiazzato tutti i genitori. In questo modo la dimensione della popolazione non è mai variata. Infine, per ciascun nuovo individuo, si sono estratti quattro numeri casuali ($q_1,q_2,q_3,q_4$), uniformemente distribuiti tra zero ed uno. Se per uno (o più) di questi $q_i<p_{mut \space i-esima}$, al commesso viene applicata la mutazione corrispondente. Le stesse operazioni descritte in questo paragrafo vengono poi ripetute per un totale di $n_{gen}=20000$. Successivamente verrà motivata la scelta di un numero così elevato.

Nei grafici successivi viene mostrato, al variare delle generazioni, il valore medio di $L^{(2)}$ per la metà della popolazione avente il percorso più breve. Viene anche disegnato il percorso dell'individuo migliore al termine della simulazione.

**NOTA:** gli assi di tutti i grafici non hanno alcuna unità di misura. La posizione delle città è stata scelta in unità arbitrarie in modo tale che le distanze in gioco non fossero né troppo grandi né troppo piccole. Se il lettore lo desidera può quindi immaginare che, nel disegno del migliore itinerario, le unità siano metri, o chilometri, o miglia (o altre). Analogo discorso vale per la $L^{(2)}$, l'unica differenza è che si sta misurando una lunghezza al quadrato.

Ciò non deve sorprendere: dei cinquanta commessi utilizzati per il calcolo di $<L^{(2)}>$, alcuni di loro potrebbero aver subito delle mutazioni sfavorevoli che ne hanno allungato il percorso rispetto agli individui migliori di quella generazione.  

Ci si può ritenere soddisfatti dei risultati ottenuti: entrambi i tragitti, sebbene non siano i più corti possibili, sono piuttosto ragionevoli e mostrano che la popolazione si è evoluta nella "direzione" desiderata.
