### **Scopo dell'esercizio è testare il generatore di numeri pseudocasuali**

- ### Verificare che $\langle r \rangle = \int_0^1 r dr = \frac{1}{2}$

Si vuole dunque verificare che il valore medio dei numeri prodotti dal generatore di numeri casuali sia $\frac{1}{2}$, come per una distribuzione di probabilità uniforme su $[0;1)$. 

per poter svolgere l'esercizio in modo ottimale è neccesario abbinare a ciascun risultato le relative incertezze. In questo e nei successivi esercizi si utilizza il metodo a blocchi. In una simulazione Monte Carlo, si divide il numero totale $M$ dei "passi" in $N$ blocchi. In ciascun blocco, si ottiene una stima $A_i$ della quantità che si intende valutare. Tramite gli  $N$ $A_i$ si possono calcolare le quantità: $$ \langle A^2 \rangle = \frac{1}{N} \sum_{i=1}^N A^2_i \quad \quad \langle A\rangle^2 = \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 $$ da cui si ottiene immediatamente la deviazione standard sulla media di A: $$\frac{\sigma}{\sqrt{N-1}} =\sqrt{\frac{1}{N-1} \left[ \langle A^2 \rangle -  \langle A\rangle^2 \right]} $$

In questo esercizio si è scelto un numero di "passi" (o "lanci" o "tiri") Monte Carlo pari a $M=10^4$ e $N=100$ blocchi. Di conseguenza, in ogni blocco, vengono effettuati $L=M/N=100$ "passi", che in questo caso corrispondono alla generazione di un $r\in[0;1)$. 

Sono stati generati, quindi, $10^4$ numeri casuali compresi tra 0 ed 1. Ogni valore di $\langle r \rangle_i$ è stato calcolato facendo la media degli $L$ numeri presenti in ciascun blocco. A questo punto, è possibile equiparare ciascun $\langle r \rangle_i$ con una misura sperimentale della quantità $\langle r \rangle$.

Viene poi graficato il valore di $\langle r \rangle$, con la relativa incertezza, in funzione del numero dei tiri totali. Inizialmente, infatti, viene considerato soltanto il primo blocco: il valore di $\langle r \rangle$ risulta pari, banalmente, ad $\langle r \rangle_1$, mentre la sua incertezza è nulla. Non ha senso infatti, in un qualunque esperimento, definire l'incertezza di una grandezza se la misura non viene ripetuta almeno due volte. Successivamente, vengono considerati sempre più blocchi: il valore di $\langle r \rangle$ risulta ora essere la media delle $\langle r \rangle_i$, l'incertezza si può calcolare tramite la formula precedente. Ovviamente il valore totale dei tiri effettuati è semplicemente $LN$, con $N$ numero dei blocchi considerati. 

I risultati ottenuti sono stati i seguenti:


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

throws, r,err = np.loadtxt("media.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(throws,r,yerr=err)
plt.title ('VALORE MEDIO DI r')
plt.xlabel('numero di lanci')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

#### Osservazioni:
Si è scelto di graficare, in funzione del numero dei tiri, $\langle r \rangle -\frac{1}{2}$ in modo da poter osservare la convergenza di tale valore verso lo zero, più facilmente visibile e identificabile sul grafico.

Come si può facilmente notare, il valore ottenuto è, entro una ${\sigma}$, compatibile con zero, ovvero quanto previsto per una distribuzione uniforme.

Si può quindi affermare che il generatore di numeri casuali abbia superato questo primo test: dopo le oscillazioni iniziali, il valore medio  $\langle r \rangle$ sembra convergere verso $\frac{1}{2}$ in modo piuttosto soddisfacente e senza mai distanziarsene per più di $2{\sigma}$.

Ovviamente, per studiare ancora più in dettaglio il comportamento $\langle r \rangle$, si potrebbe aumentare il numero totale dei tiri. Se il numero dei blocchi non varia, ciascun $\langle r \rangle_i$ sarebbe ottenuto dalla media di un numero maggiore di numeri casuali. Di conseguenza, le incertezze risulterebbero ridotte. 

- ### Verificare che $\sigma^2 = \int_0^1 (r-\frac{1}{2})^2 dr = \frac{1}{12}$

Si vuole dunque verificare che la varianza coincida con quella di una variabile distribuita uniformemente tra $[0;1)$.

Questo esercizio viene risolto con un procedimento analogo a quanto precedentemente descritto: non vengono variati i valori di $M$ e di $N$ e, per fare tutte le analisi dati, vengono utilizzati gli stessi $10^4$ numeri casuali con cui si era studiato l'andamento del valor medio.

L'unica differenza risiede, ovviamente, nel calcolo degli $A_i$: la media viene effettuata tra i $(r-1/2)^2$ (che sono $L$ per ciascun blocco), dal momento che si sta studiando la varianza e non più $\langle r \rangle$. 

Viene poi graficato il valore di $\langle(r-\frac{1}{2})^2\rangle-\frac{1}{12}$ in funzione del numero dei blocchi considerati. È stato sottratto $\frac{1}{12}$ per poter osservare la convergenza al valore zero.

```python
throws, r,err = np.loadtxt("varianza.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(throws,r,yerr=err)
plt.title('VARIANZA')
plt.xlabel('numero di lanci')
plt.ylabel('<(r-1/2)^2>-1/12')
plt.grid(True)
plt.show()
```

#### Osservazioni:
Anche in questo caso, i valori ottenuti non si discostano mai eccessivamente dal valore atteso e, considerando tutti i 100 blocchi, il valore ottenuto è estremamente vicino a zero.
Si può dunque affermare che il generatore di numeri casuali utilizzato abbia superato anche questo secondo test. 

- ### Effettuare il test del $\chi^2$ 100 volte dividendo l'intervallo $[0;1)$ in 100 sottointervalli uguali

Il test del $\chi^2$ è una misura dell'errore tra le osservazioni $O_i$ e i valori attesi $E_i$. La formula utilizzata è la seguente: $$\chi^2 = \sum_{i} \frac{\left( O_i - E_i \right)^2}{E_i}$$ 

In questo specifico caso, se l'intervallo $[0;1)$ viene diviso in $M=100$ sottointervalli, la probabilità che un numero appartenga ad uno specifico sottointervallo è pari a $p=\frac{1}{M}$ (nell'ipotesi che la distribuzione di probabilità del generatore sia uniforme tra 0 ed 1). Poiché $p<<1$ è possibile utilizzare la formula approssimata del $\chi^2$ scritta sopra: al denominatore dovrebbe esserci la varianza secondo la definizione, tuttavia $\sigma^2$ è circa uguale al valore atteso per piccole probabilità.

Per ogni sottointervallo, il numero di occorrenze attese è $E_i=n/M$. Chiamando $n_i$ il numero di occorrenze osservate per ciascun sottointervallo, la formula precedente diventa dunque:
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$

In media,se la distribuzione fosse uniforme, ci si dovrebbe attendere che $(n_i-n/M)^2 \simeq n/M$ e, quindi, dall'equazione precedente, $\chi^2 \simeq M$. Se il risultato del test fosse molto maggiore di $M$ si dovrebbe rigettare l'ipotesi che i numeri generati siano uniformemente distribuiti. 

Spesso, al posto del $\chi^2$ si preferisce studiare il $\chi^2_{ridotto}=\frac{\chi^2}{NDF}$, con $NDF$ numero dei gradi di libertà (in questo caso $NDF=99$, 100 sottointervalli meno il vincolo di avere $10^4$ numeri). Dal valore del $\chi^2_{ridotto}$ si trovano poi tabulati i livelli di confidenza $CL$ con cui è possibile rigettare l'ipotesi. I $CL$ dipendono in modo complicato sia dal valore del $\chi^2_{ridotto}$ che dai $NDF$, tuttavia si può affermare che, in generale, se $\chi^2_{ridotto}<2$ non è possibile rigettare l'ipotesi con un $CL$ ragionevolmente alto.

Si decide di ripetere il test 100 volte, avendo cura, ogni volta, di rigenerare i $10^4$ numeri casuali. In questo modo si può verificare se la distribuzione è uniforme in tutta la sequenza dei numeri casuali, o solo in alcune parti.

Dopo aver implementato il codice per il calcolo del $\chi^2$, i risultati ottenuti sono stati i seguenti:


```python
j, chi = np.loadtxt("chiquadro.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(j,chi)
plt.title('Test $\chi^2$')
plt.xlabel('Numero test')
plt.ylabel('$\chi^2$')
plt.grid(True)
plt.show()
```

#### Osservazioni:
Come si può osservare dal grafico, il valore del $\chi^2$ oscilla intorno al valore $100$ per tutti i test: non si hanno mai valori pari o superiori a 140.
Non è quindi possibile rigettare l'ipotesi che il generatore produca numeri casuali distribuiti uniformemente in ogni parte della sua sequenza.
Si può dunque affermare che anche questo terzo test statistico sia stato superato. 
Avere la certezza che il generatore produca effettivamente numeri casuali uniformemente distribuiti tra 0 ed 1 risulterà fondamentale nel prossimo esercizio in cui si cercherà di produrre numeri secondo densità di probabilità diverse (ad esempio esponenziale o lorentziana).

## <span style="color:blue">ESERCIZIO 1.2</span>

### **Scopo dell'esercizio è aggiungere due distribuzioni di probabilità al generatore di numeri casuali e di verificare il teorema del limite centrale**

- ### Aggiungere al generatore di numeri casuali la distribuzione di probabilità esponenziale $p(x) = \lambda \exp(-\lambda x)$ con $x\in [0;+\infty]$ e quella lorenziana $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$ con $x\in [-\infty;+\infty]$.
Sia $p_x(x)$ una distribuzione di probabilità generica e $F_x(x)=\int_{-\infty}^x p(t)dt$ la sua funzione cumulativa. Se $y=y(x)$ è una funzione continua e monotona in x vale che $p_y(y)=p_x(x) \left\lvert{\frac{dy(x)}{dx}}\right\rvert^{-1}$. Se viene scelta $y(x)=F_x(x)$, risulta che $p_y(y)=1$ con $y \in [0;1)$.

Di conseguenza, se la $F_x(x)$ è nota ed è invertibile analiticamente, si ha che $x=F_x^{-1}(y)$ sono numeri casuali con distribuzione di probabilità data da $p_x(x)$. Si comprende quindi, concretamente, perché è cruciale avere un generatore di numeri (pseudo-)casuali distribuiti uniformemente tra 0 e 1 (e il motivo di tutte le verifiche dell'esercizio precedente).  

La funzione cumulativa della distribuzione esponenziale si ottiene facilmente integrando la $p(x)$ da zero a più infinito (l'estremo iniziale di integrazione non è meno infinito perché la distribuzione è definita solo su $\mathbb{R}^+$). Si ottiene che $F(x)=1-e^{-{\lambda}x}$, funzione facilmente invertibile analiticamente. Dalle formule precedentemente esposte, si comprende che, se y è uniformemente distribuita tra 0 e 1, allora: $$x=-\frac{1}{\lambda}ln\left(1-y\right)$$ è distribuita esponenzialmente. 

Una prima banale verifica può essere effettuata osservando che l'argomento del logaritmo è sempre minore di uno, pertanto x sarà un numero necessariamente positivo (a causa del segno negativo davanti alla formula), come deve necessariamente essere per il dominio della $p(x)$.

Un analogo ragionamento si può ripetere anche per la distribuzione lorentziana (l'unica differenza è che il dominio di $p(x)$ è $\mathbb{R}$. La funzione cumulativa risulta essere $F(x)=\frac{1}{\pi}arctan\left(\frac{x-\mu}{\Gamma}\right)+\frac{1}{2}$. Anche questa funzione si può invertire analiticamente e, in definitiva, si ha che: $$x=\mu+{\Gamma}tan\left[{\pi}\left(y-\frac{1}{2}\right)\right]$$

- ### Verifica del teorema del limite centrale
<span style="color:red"> **TEOREMA DEL LIMITE CENTRALE:** </span> Date $N$ variabili **indipendenti** e **identicamente distribuite**, ovvero $\langle x_1\rangle=\dots=\langle x_N\rangle=\mu$ e $\langle x_1^2\rangle-\langle x_1\rangle^2=\dots=\langle x_N^2\rangle-\langle x_N\rangle^2=\sigma^2$ a **varianza finita** $\sigma^2<+\infty$, allora, se $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, vale che: $$\langle S_N\rangle=\mu \quad \quad \sigma^2_{S_N}=\frac{\sigma^2}{N}$$ Inoltre la **distribuzione di probabilità, per $N\to+\infty$, tende a una gaussiana** con valore medio $\mu$ e varianza $\sigma^2_{S_N}$.

Per verificare questo teorema si utilizza, inizialmente, un dado classico. Il risultato di ogni singola estrazione può essere un qualunque intero tra 1 e 6. Ovviamente, si assume che ciascun numero abbia la stessa probabilità di essere estratto.

Per implementare questa distribuzione discreta sul calcolatore, si generano valori casuali $x$ uniformemente distribuiti nell'intervallo $[1;7)$. Per avere solo numeri interi è poi sufficiente utilizzare il comando $(int)x$, che arrotonda per difetto qualunque numero con la virgola.

Vengono successivamente generati $10^4$ valori di $S_N$ (con $N=1,2,10,100$). Per $N=1$ la distribuzione di $S_N$ coincide banalmente con quella uniforme, quindi può essere considerata un'ulteriore verifica della bontà del generatore di numeri casuali. All'aumentare di N, per il teorema del limite centrale, il valore medio non dovrebbe variare da $\mu=3,5$ in questo caso, mentre la distribuzione di probabilità dovrebbe assomigliare sempre di più ad una gaussiana.

Per visualizzare i dati, si possono disegnare degli istogrammi per ognuno dei quattro valori di N. Il range e il numero di bin sono stati scelti a seguito di diversi tentativi, l'obiettivo è infatti quello di fornire grafici che siano di più semplice lettura possibile.

```python
from scipy.optimize import curve_fit

def f(x,var,mu,k):  
    return k/(np.sqrt(2*np.pi*var)) * np.exp(-((x-mu)**2)/(2*var))

data = np.loadtxt("uniforme1.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 6, range=(0.5,6.5),ec='grey')
plt.title('Distribuzione uniforme $N=1$')
plt.xlabel('$S_1$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

data = np.loadtxt("uniforme2.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 12, range=(0.5,6.5),color='green',ec='grey')
plt.title('Distribuzione uniforme $N=2$')
plt.xlabel('$S_2$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

data = np.loadtxt("uniforme10.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 12, range=(1.5,5),color='purple',ec='grey')
plt.title('Distribuzione uniforme $N=10$')
plt.xlabel('$S_{10}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

data = np.loadtxt("uniforme100.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(3.,4.),color='red',ec='grey')
data_entries, bins = np.histogram(data, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
N = 100
x = np.linspace(3.,4.,N,endpoint=True)
p_opt, p_cov = curve_fit(f, xdata=binscenters, ydata=data_entries)
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit) 

plt.title('Distribuzione uniforme $N=100$')
plt.xlabel('$S_{100}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

print("Parametri di best-fit per la Gaussiana [varianza,mu,k]=")
print(p_opt)
print("Incertezza sui parametri=")
print(np.sqrt(np.diagonal(p_cov)))

E_i=f(binscenters,p_opt[0],p_opt[1],p_opt[2])
sigma_chiquadro=np.sqrt(E_i)
chiquadro_i=((data_entries-E_i)/(sigma_chiquadro))**2

chiquadro=np.sum(chiquadro_i)
print('chi-quadro=')
print(chiquadro)

```

Come ci si poteva attendere, per $N=1$ tutti i valori sono usciti un numero molto simile di volte.

Per $N=2$ ed $N=10$ la distribuzione di probabilità di $S_N$ comincia ad assomigliare ad una gaussiana: le occorrenze intorno al valore medio della distribuzione ($\mu=3,5$) aumentano sensibilmente, a scapito di quelle dei valori estremi. L'istogramma di $N=2$ sembra avere un problema: più di $200$ il risultato dell'estrazione è maggiore di 6 (ciò è assurdo perché $\forall N$ $S_N \in [1;6]$). In realtà si tratta solamente di una convenzione nel riempimento dei bin. In un qualunque bin che ha estremi $x_{min}$ e $x_{max}$, viene infatti inserito un determinato dato $x$ se $x \in [x_{min};x_{max})$. Di conseguenza, tutte le $S_2=6$ vengono inserite in un intervallo che sembra oltre il dominio di definizione di $S_N$.

Per $N=100$ la somiglianza della curva con una gaussiana è molto forte, pertanto si è effettuato un fit con una generica curva del tipo $$f(x)=\frac{k}{2{\pi}\sigma^2}exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$ Ovviamente $\mu$ è il valore medio e $\sigma^2$ la varianza. $k$ è invece un parametro di (non-)normalizzazione. Nella distribuzione di probabilità gaussiana $k=1$, tuttavia quella ottenuta dai dati sperimentali certamente non è normalizzata e necessita, quindi, di questo parametro aggiuntivo. Questo parametro non è però interessante dal punto di vista statistico, quindi verrà trascurato in tutte le successive discussioni.

I parametri di questa curva confermano, grosso modo, le tesi del teorema del limite centrale. Il valore medio della distribuzione $S_N$ dovrebbe rimanere pari a $\mu$, indipendentemente dal valore di $N$. Il $\mu_{best}$ è molto vicino a $3,5$ (differisce di $0.004$). Tuttavia la sua incertezza è solo 0.002 e, quindi, ci si trova ad una distanza di circa $2\sigma$ dal valore atteso. Questo non deve sorprendere eccessivamente, il risultato è quasi certamente dovuto ad una fluttuazione statistica: si sono generati "solamente" $10^4$ valori di $S_{100}$, il risultato tornerebbe compatibile, entro le incertezze, con il valore di $3,5$ se si aumentasse, di qualche ordine di grandezza, il numero delle $S_{100}$. Bisogna ricordarsi, inoltre, che esistono vari modi per calcolare i coefficienti (e le loro incertezze) di una curva di best-fit e che ciascuno di questi produce un risultato leggermente diverso.

La varianza di una distribuzione uniforme discreta è pari a $\sigma^2=\frac{n^2-1}{12}$, quindi, in questo caso, $\frac{35}{12}$. Per il teorema del limite centrale, $\sigma^2_{S_{100}}=\frac{35}{1200}\simeq 0,0292$. Si nota che questo valore è compatibile con quello ottenuto dal fit entro le incertezze.

Per un'analisi più approfondita si decide anche di effettuare un test del $\chi^2$. In questo caso, i valori osservati sono pari al numero delle occorrenze di ogni bin. Il valore atteso si ottiene invece calcolando l'ordinata della gaussiana di best-fit al centro di quel bin. Ancora una volta è possibile approssimare la varianza al denominatore della formula del $\chi^2$ con il valore atteso. Risulta pertanto che: $$\chi^2 = \sum_{i} \frac{\left( O_i - E_i \right)^2}{E_i}=\sum_{i} \left(\frac{N_{occ_i}-f(\langle x \rangle_{bin_i})}{\sigma^2_i}\right)$$ Si ottiene un valore di circa 26. I gradi di libertà sono 17 (il numero di bin meno i tre vincoli imposti dalla curva di best-fit), pertanto il $\chi^2_{ridotto}$ è leggermente superiore a $1,5$. Non è quindi possibile rigettare l'ipotesi, con un livello di confidenza sufficientemente alto, che i dati dell'istogramma corrispondano effettivamente a una gaussiana definita dai parametri precedentemente calcolati.

Si può dunque affermare che il teorema del limite centrale sia stato verificato in questo primo caso.

Nel secondo caso si ripete la stessa simulazione utilizzando un dado esponenziale con $\lambda=1$. Tuttavia, i valori generati appartengono ad un intervallo illimitato e non vengono più arrotondati all'intero precedente: la distribuzione esponenziale è definita su un intervallo continuo.

```python
data = np.loadtxt("exp1.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 12, range=(0,6),ec='grey')
plt.title('Distribuzione esponenziale $N=1$')
plt.xlabel('$S_1$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()
data = np.loadtxt("exp2.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 12, range=(0,6),color='green',ec='grey')
plt.title('Distribuzione esponenziale $N=2$')
plt.xlabel('$S_2$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()
data = np.loadtxt("exp10.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 15, range=(0,2.5),color='purple',ec='grey')
plt.title('Distribuzione esponenziale $N=10$')
plt.xlabel('$S_{10}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

data = np.loadtxt("exp100.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(0.5,1.5),color='red',ec='grey')
data_entries, bins = np.histogram(data, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
N = 100
x = np.linspace(0.5,1.5,N,endpoint=True)
p_opt, p_cov = curve_fit(f, xdata=binscenters, ydata=data_entries, bounds=([0,-1,0],[1,1,1000]))
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit) 

plt.title('Distribuzione esponenziale $N=100$')
plt.xlabel('$S_{100}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

print("Parametri di best-fit per la Gaussiana [varianza,mu,k]=")
print(p_opt)
print("Incertezza sui parametri=")
print(np.sqrt(np.diagonal(p_cov)))

E_i=f(binscenters,p_opt[0],p_opt[1],p_opt[2])
sigma_chiquadro=np.sqrt(E_i)
chiquadro_i=((data_entries-E_i)/(sigma_chiquadro))**2

chiquadro=np.sum(chiquadro_i)
print("chi-quadro:")
print(chiquadro)
```

Per $N=1$ la distribuzione di $S_1$ coincide con quella esponenziale semplice attesa teoricamente: il primo bin dell'istogramma risulta essere quello più riempito, successivamente vi è un rapido decadimento.

Per $N=2$ e $N=10$ la distribuzione comincia a essere più piccata in corrispondenza del valore medio $\lambda$, mentre, per $N=100$, abbiamo una forte somiglianza con una curva gaussiana.

Effettuando il fit valgono, approssimativamente, tutte le considerazioni precendenti: la varianza ottenuta è compatibile, entro la sua incertezza, con il valore predetto dal teorema del limite centrale. $\sigma^2_{S_{100}}=\frac{\sigma^2}{100}$, con $\sigma^2=\frac{1}{\lambda^2}=1$ varianza della distribuzione esponenziale. Il valore medio, invece, pur essendo molto vicino a quello previsto, ne è compatibile solamente entro $4\sigma$. Ancora una volta, ciò può essere spiegato dalle fluttuazioni statistiche (non si sono generati infiniti $S_{100}$).

Il valore del $\chi^2$ risulta essere notevolmente più grande rispetto al caso precedente (a fronte dello stesso numero di gradi di libertà). Uno studio più approfondito mostra però che più della metà del suo valore è determinato dai bin più esterni. Viene allora ripetuto lo studio escludendo quei valori estremi:

```python
data = np.loadtxt("exp100.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(0.8,1.2),color='red',ec='grey')
data_entries, bins = np.histogram(data, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
N = 100
x = np.linspace(0.8,1.2,N,endpoint=True)
p_opt, p_cov = curve_fit(f, xdata=binscenters, ydata=data_entries,bounds=([0,-1,0],[1,1,1000]))
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit)
plt.title('Distribuzione esponenziale $N=100$') 
plt.xlabel('$S_{100}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

print("Parametri di best-fit per la Gaussiana [varianza,mu,k]=")
print(p_opt)
print("Incertezza sui parametri=")
print(np.sqrt(np.diagonal(p_cov)))

E_i=f(binscenters,p_opt[0],p_opt[1],p_opt[2])
sigma_chiquadro=np.sqrt(E_i)
chiquadro_i=((data_entries-E_i)/(sigma_chiquadro))**2

chiquadro=np.sum(chiquadro_i)
print("chi-quadro:")
print(chiquadro)
```

Come si nota immediatamente, i parametri di best fit non sono variati sensibilmente, mentre il $\chi^2$ si è ridotto di molto: il $\chi^2_{ridotto}$ risulta essere circa 1. I dati centrali si adattano quindi estremamente meglio alla curva gaussiana.

Questa enorme differenza con l'istogramma precedente è facilmente spiegabile: i bin più esterni si trovano a varie $\sigma$ dal valore medio, hanno quindi una probabilità molto bassa di essere occupati. Per questo motivo, una differenza tra il valore atteso e osservato anche di poche unità (derivante dalle fluttuazioni statistiche), può determinare un enorme contributo al $\chi^2$.

Sulla base di queste considerazioni, si può quindi affermare che il teorema del limite centrale risulta verificato anche in questo secondo caso.
Si ripete una terza volta lo stesso esperimento, utilizzando però un dado lorentziano (con $\mu=0$ e $\Gamma=1$).

```python
data = np.loadtxt("lor1.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(-10,10),ec='grey')
plt.title('Distribuzione lorentziana $N=1$')
plt.xlabel('$S_1$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()
data = np.loadtxt("lor2.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(-10,10),color='green',ec='grey')
plt.title('Distribuzione lorentziana $N=2$')
plt.xlabel('$S_2$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()
data = np.loadtxt("lor10.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(-10,10),color='purple',ec='grey')
plt.title('Distribuzione lorentziana $N=10$')
plt.xlabel('$S_{10}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

data = np.loadtxt("lor100.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(-10,10),color='red',ec='grey')
data_entries, bins = np.histogram(data, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
N = 100
x = np.linspace(-10,10,N,endpoint=True)
p_opt, p_cov = curve_fit(f, xdata=binscenters, ydata=data_entries,bounds=([1,-1,7000],[2,1,8000]))
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit) 
plt.title('Distribuzione lorentziana $N=100$')
plt.xlabel('$S_{100}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

print("Parametri di best-fit per la Gaussiana [varianza,mu,k]=")
print(p_opt)
print("Incertezza sui parametri=")
print(np.sqrt(np.diagonal(p_cov)))

E_i=f(binscenters,p_opt[0],p_opt[1],p_opt[2])
sigma_chiquadro=np.sqrt(E_i)
chiquadro_i=((data_entries-E_i)/(sigma_chiquadro))**2

chiquadro=np.sum(chiquadro_i)
print("chi-quadro:")
print(chiquadro)
```

Rispetto ai due esperimenti precedenti ci sono delle notevolissime differenze: le $S_N$, innanzitutto, non sembrano cambiare forma all'aumentare di N. Inoltre per $N=100$ il fit con la gaussiana fornisce dei risultati totalmente scadenti: il $\chi^2$ è elavatissimo e, anche visivamente, la gaussiana si discosta completamente dall'istogramma appena ci si allontana dal valor medio. 

La spiegazione di questi risultati si trova nelle ipotesi del teorema del limite centrale: la distribuzione di probabilità di partenza deve avere varianza finita. La lorentziana, tuttavia, non soddisfa questa richiesta: si nota immediatamente che l'integrale da risolvere per calcolarne il valor medio diverge, dunque la $\sigma^2$ non è definita.

Per concludere questo esercizio, si decide di verificare anche la generalizzazione del teorema del limite centrale: se le $N$ variabili indipendenti **NON** sono identicamente distribuite (ma comunque vale che $\sigma_i<+\infty$), allora $\langle S_N \rangle$ tende alla media delle medie, mentre la varianza tende alla media delle varianze diviso N. La distribuzione di probabilità di $S_N$ tenderà ancora ad una gaussiana definita dalla $\langle S_N \rangle$ e dalla $\sigma^2_{S_N}$.

Per verificare questa generalizzazione si sommano tra loro $N=200$ variabili indipendenti, cento delle quali sono distribuite uniformemente in $[0;10)$, le altre distribuite esponenzialmente con $\lambda=1$. Anche in questo caso, si calcola $10^4$ volte $S_{200}$ e analizzo i dati ottenuti tramite un istogramma.

```python
data = np.loadtxt("somme.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 20, range=(2.5,3.5),color='red',ec='grey')
data_entries, bins = np.histogram(data, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
N = 100
x = np.linspace(2.5,3.5,N,endpoint=True)
p_opt, p_cov = curve_fit(f, xdata=binscenters, ydata=data_entries)
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit) 
plt.title('Distribuzione $N=200$')
plt.xlabel('$S_{200}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()


print("Parametri di best-fit per la Gaussiana [varianza,mu,k]=")
print(p_opt)
print("Incertezza sui parametri=")
print(np.sqrt(np.diagonal(p_cov)))

E_i=f(binscenters,p_opt[0],p_opt[1],p_opt[2])
sigma_chiquadro=np.sqrt(E_i)
chiquadro_i=((data_entries-E_i)/(sigma_chiquadro))**2

chiquadro=np.sum(chiquadro_i)
print("chi-quadro:")
print(chiquadro)
```

In questo caso, poiché le varianze sono tutte finite, tornano a essere valide le ipotesi, e di conseguenza anche le tesi, del teorema del limite centrale. In questo caso, inoltre, sia il valore medio che la varianza sono compatibili, entro le loro incertezze con i valori attesi: $\langle S_{200}\rangle=\frac{5+1}{2}=3$ e $\sigma_{S_{200}}=\frac{\frac{1+100/12}{2}}{200}\simeq0,0233$.
Anche il $\chi^2$ è ottimo: una volta diviso per il numero dei gradi di libertà si ottiene un valore di poco superiore ad uno.

Infine, un'ulteriore generalizzazione del teorema può essere effettuata rimuovendo la richiesta che le N variabili siano indipendenti. Il teorema continua a valere, a patto che ciascuna variabile sia correlata con un numero finito di altre variabili che definiscono $S_N$.

Per effettuare questa verifica, si generano $10^4$ valori di $S_{200}$, definita come somma di 100 variabili indipendenti distribuite esponenzialmente con $\lambda=1$, e altri cento numeri pari, ciascuno, al doppio di uno dei primi cento. In questo modo, risulta che ciascun addendo è correlato solamente ad un altro, dunque dovrebbero essere soddisfatte le ipotesi del teorema.

```python
data = np.loadtxt("corr.dat", usecols=(0), delimiter=' ', unpack='true')
n, bins, patches = plt.hist(data, 16, range=(1.2,1.8),color='red',ec='grey')
data_entries, bins = np.histogram(data, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
N = 100
x = np.linspace(1.2,1.8,N,endpoint=True)
p_opt, p_cov = curve_fit(f, xdata=binscenters, ydata=data_entries,bounds=([0,1,300],[1,2,400]))
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit) 
plt.title('Distribuzione $N=200$')
plt.xlabel('$S_{200}$')
plt.ylabel('occorrenze')
plt.grid(True)
plt.show()

print("Parametri di best-fit per la Gaussiana [varianza,mu,k]=")
print(p_opt)
print("Incertezza sui parametri=")
print(np.sqrt(np.diagonal(p_cov)))

E_i=f(binscenters,p_opt[0],p_opt[1],p_opt[2])
sigma_chiquadro=np.sqrt(E_i)
chiquadro_i=((data_entries-E_i)/(sigma_chiquadro))**2

chiquadro=np.sum(chiquadro_i)
print("chi-quadro:")
print(chiquadro)
```

Si nota dunque che, anche in questo caso, i risultati verificano il teorema del limite centrale poiché il valore del $\chi^2$ è comparabile con il numero dei gradi di libertà.

#### Brevi considerazioni finali
Quando le variabili di partenza rispettavano le ipotesi, si sono sempre verificate le tesi del teorema del limite centrale.
L'unico tra i precedenti casi, quindi, in cui la distribuzione di $S_N$ non tende ad una gaussiana è quello dei dati distribuiti secondo una curva lorentziana. Inoltre si osservava che la forma di $S_N$ non subiva variazioni significative all'aumentare di $N$. Si può dimostrare, infatti, che la lorentziana è la funzione di distribuzione limite per un'ampia classe di distribuzioni di probabilità a varianza infinita.

## <span style="color:blue">ESERCIZIO 1.3</span>
### SIMULARE L'ESPERIMENTO DI BUFFON E FORNIRE UNA STIMA DI  $\pi$  AL VARIARE DEL NUMERO DI LANCI $M$

L'esperimento di Buffon è, concettualmente, molto semplice: su un pavimento decorato con strisce equidistanti $d$, viene lanciato casualmente un ago di lunghezza $l$ (con $l<d$). Lo scopo dell'esperimento è calcolare la probabilità che l'ago intersechi una delle strisce del pavimento.

La probabilità può essere calcolata anaticamente e risulta essere: $P=\frac{2l}{d\pi}$. Risulta allora evidente come, tramite l'esperimento, si possa ottenere una stima del valore di $\pi=\frac{2l}{Pd}$. Se il numero di tiri $M_{tiri}\to+\infty$, la probabilità si può esprimere come $P=\frac{M_{hit}}{M_{tiri}}$ con $M_{hit}$ numero dei lanci in cui l'ago risulta avere intersezione con una striscia. In definitiva, tramite questo esperimento, è quindi possibile stimare il valore di $\pi$ con la formula: $$\pi=\lim_{M_{tiri} \to \infty}\frac{2LM_{tiri}}{M_{hit}d}$$

Per simulare l'esperimento è necessario usare alcuni accorgimenti. Nonostante l'esperimento possa sembrare molto semplice, presenta alcune difficoltà computazionali: innanzitutto il pavimento di Buffon sarebbe idealmente infinito, e, inoltre, si osserverà che risulta naturale utilizzare $\pi$ per stimare il valore di $\pi$ stesso. Ovviamente si dovranno utilizzare strategie per risolvere queste complicazioni. 

Per evitare di considerare un pavimento infinito, è sufficiente osservare che, ovunque cada il centro dell'ago, esisterà sempre una riga della griglia più vicina ad esso. Nel caso in cui il centro fosse esattamente equidistante a due strisce, se ne può considerare, indifferentemente, una sola. Assumendo che l'ago possa cadere in ogni posizione con la stessa probabilità, la distribuzione di probabilità della distanza del centro dell'ago dalla riga più vicina è quindi uniforme tra $x\in[0;\frac{d}{2})$. Se infatti $x>\frac{d}{2}$, la riga considerata non è quella più vicina al centro dell'ago.

Tramite l'angolo acuto $\theta$ che l'ago (o il suo prolungamento) forma con quella striscia del pavimento, si ricava immediatamente una relazione geometrica tra $\theta$ e $x$ che viene soddisfatta solo in caso di intersezione.

<img src="buffon.png">

È evidente che c'è intersezione se $x\leqslant\frac{l}{2}\sin{\theta}$. Se l'ago viene lanciato casualmente, nessun valore di $\theta$ risulta privilegiato, pertanto anche la sua distribuzione è uniforme. 

Sembra, dunque, che si sia annullata la difficoltà computazionale dell'esperimento: è sufficiente considerare un numero grande di $M_{tiri}$ (a volte verrà chiamato semplicemente $M$), generare $M$ valori uniformemente distribuiti in $[0;\frac{d}{2})$ per la posizione del centro dell'ago e altrettanti valori per $\theta$ (anch'essa uniformemente distribuita). $M_{hit}$ risulta essere pari al numero di coppie $(x;\theta)$ che soddisfano la relazione geometrica precedente.

Tuttavia, un'ulteriore difficoltà si cela nell'angolo $\theta$: la sua distribuzione è uniforme tra $[0;\frac{\pi}{2})$. Per stimare il valore di $\pi$, sembra dunque necessario dover utilizzare $\pi$.

Inizialmente, si simula l'esperimento nonostante questo problema. Il valore di $M$ viene posto pari a $10^5$, la distanza tra le strisce del pavimento è $d=1$ e la lunghezza dell'ago è $l=0,2$. Sono state poi generate $M$ coppie casuali $(x;\theta)$ secondo le distribuzioni di probabilità precedentemente discusse.

Come nell'esercizio 1.1, l'esperimento viene suddiviso in $N=100$ blocchi. In questo modo il valore di $\pi$ può essere calcolato $N$ volte ed è quindi possibile valutarne l'incertezza.
Si disegna allora un grafico del valore medio di $\pi$ in funzione del numero crescente di blocchi $N=1,2,\dots,100$ (e quindi del numero totale di tiri considerati). Questi sono stati i risultati ottenuti:

```python
i, pi, err = np.loadtxt("pi1.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,pi,yerr=err)
plt.xlabel('numero tiri')
plt.ylabel('$\pi$')
plt.grid(True)
N = 100
x = np.linspace(0.,100000.,N,endpoint=True)
y=np.zeros(N)
y=y+np.pi
plt.plot(x,y,color="red") 
plt.show()
```

Si nota immediatamente che il valore ottenuto con tutti i $10^5$ tiri è in ottimo accordo con il valore tabulato di $\pi$ (disegnato in rosso). Analizzando l'intero grafico, si può affermare che la convergenza sembra buona: dopo le oscillazioni iniziali, i risultati sono sempre compatibili, entro le loro incertezze, con il valore atteso.

Tuttavia, nonostante un elevato numero di lanci totali, le incertezze rimangono piuttosto elevate (quella corrispondente a $10^5$ tiri è di circa $0,025$). Per diminuirle vi sono due possibilità: aumentare ulteriormente il numero $M$, oppure allungare l'ago (mantendendo sempre però $l<d$). 

Si è scelta quest'ultima strada: se $d\simeq l$ si riducono notevolmente le possibili fluttuazioni statistiche di $M_{hit}$, i valori di $\pi$ calcolati dai vari blocchi risultano quindi più simili tra loro e, di conseguenza, l'incertezza diminuisce.

Si è quindi posto $l=0,95$ e, con le stesse coppie casuali precedenti, si sono ricalcolati i valori di $\pi$ dei vari blocchi con le formule derivanti dalle condizioni geometriche.
Viene disegnato quindi un grafico analogo al precedente:

```python
i, pi, err = np.loadtxt("pi1_cr.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,pi,yerr=err)
plt.xlabel('numero tiri')
plt.ylabel('$\pi$')
plt.grid(True)
plt.plot(x,y,color="red") 
plt.show()
```

Potrebbe sembrare, a prima vista, che nulla sia cambiato rispetto al grafico precedente: valgono le medesime considerazioni sulla convergenza dei risultati all'aumentare del numero dei lanci considerati. Tuttavia vi è una differenza notevole, l'ultimo dato ha un'incertezza di circa $0,007$ (più di tre volte inferiore rispetto alla simulazione precedente). Come ci si poteva attendere, quindi, più le dimensioni dell'ago e la distanza delle strisce diventano comparabili, minore sarà l'incertezza sui risultati finali.

L'obiettivo è, infine, quello di simulare l'esperimento senza mai dover ricorrere all'utilizzo del valore tabulato di $\pi$ che era stato utilizzato per poter generare una direzione casuale dell'ago rispetto alla riga del pavimento.

Serve, quindi, una modalità per poter generare direzioni casuali senza dover ricorrere a $\pi$. Una soluzione è la seguente: vengono generati due numeri casuali, entrambi distribuiti uniformemente in $[-1;1)$. I due numeri possono essere considerati come le coordinate cartesiane bidimensionali di un punto. Il punto viene rigettato se non appartiene al cerchio di raggio $1$ inscritto nel quadrato $[-1;1]\times[-1;1]$. Se vi appartiene, invece, si è generata la direzione casuale desiderata (identificata dal vettore che congiunge il centro della circonferenza con il punto). Con questa tecnica ogni direzione è evidentemente equiprobabile (in un circonferenza non esiste un vettore privilegiato).

<img src="vettore.png">

Avendo a disposizione la direzione casuale, si può facilmente utilizzare la relazione geometrica precedente che determina la condizione affinché vi sia intersezione. Il seno dell'angolo acuto, infatti, tra la striscia verticale e l'ago si calcola tramite: $\sin{\theta}=\frac{|x|}{\sqrt{x^2+y^2}}$, in cui x e y sono le coordinate del punto all'interno della circonferenza.

Per simulare l'esperimento si continua ad utilizzare $l=0.95$ e $d=1$. Vengono invece aumentati a $M=10^7$ il numeri di tiri totali (a fronte di un numero di blocchi fisso a $N=100$). In questo modo si dovrebbero ottenere risultati con un'incertezza ancora più piccola (a scapito di un maggior tempo computazionale).

Vengono quindi generati $10^7$ valori uniformemente distribuiti tra $[0;\frac{d}{2})$ che rappresentano, come nel caso precedente, la distanza del centro dell'ago dalla riga più vicina. Le $10^7$ direzioni vengono invece generate come appena spiegato (se il punto non appartiene alla circonferenza viene rigettato e ovviamente non viene conteggiato).

Viene infine disegnato, ancora una volta, il grafico di $\pi$ in funzione del numero dei tiri considerati (bisogna notare che in questo caso ciascun blocco sarà costituito da un numero cento volte maggiore di "passi" rispetto ai due casi precedenti).

```python
i, pi, err = np.loadtxt("pi2.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,pi,yerr=err)
plt.xlabel('numero tiri')
plt.ylabel('$\pi$')
plt.grid(True)
x = np.linspace(0.,10000000.,N,endpoint=True)
plt.plot(x,y,color="red") 
plt.show()
```

Si osserva che, nonostante l'elevato numero dei tiri totali, il programma in C++ con cui si è effettuata la simulazione ha un tempo di esecuzione che è di circa un secondo (quindi del tutto accettabile).

Come previsto, aumentando $M$ si ottiene un risultato molto più preciso: l'incertezza dell'ultimo dato graficato è, circa, $8,5\times10^{-4}$, quindi più piccola di quasi un ordine di grandezza rispetto alla simulazione precedente.

Infine, tutti i dati riportati nel grafico sono compatibili, entro $2\sigma$ con il valore tabulato. Si può quindi concludere che la tecnica utilizzata per produrre direzioni casuali senza utilizzare $\pi$ sia efficace e, di conseguenza, l'esperimento sia riuscito anche in quest'ultimo caso.
