## <span style="color:blue">ESERCIZIO 10.1</span>

**NOTA:** in tutta la spiegazione dell'esercizio alla temperatura non verrà mai assegnata un'unità di misura, in quanto fittizia. In ogni caso l'algoritmo si basa sulla meccanica statistica in cui le temperature sono sempre espresse in kelvin. 

Per implementare il simulated annealing si è sfruttata, con le dovute modifiche, la stessa classe "popolazione" della precedente esercitazione. Innanzitutto, a differenza dell'algoritmo genetico, non si ha più a che fare con un insieme di individui, ma solamente con un unico commesso viaggiatore.

La codifica del simulated annealing si è rivelata piuttosto immediata (è stato sufficiente aggiungere una semplice funzione alla classe), dal momento che è del tutto analoga a quella dell'algoritmo di Metropolis. È stato invece più complesso definire un intervallo ragionevole di temperature fittizie, il numero di step per ciascuna temperatura e come proporre la configurazione successiva.

Si è deciso, dopo vari tentativi, di partire da una temperatura di $100$ e di terminare a $0,01$. Queste scelte sono sembrate piuttosto ragionevoli: ad una $T$ ancora superiore tutte le mosse verrebbero accettate poiché $e^x\to 1$ per $x\to 0$ ($\beta=\frac{1}{T}$ risulterebbe molto piccolo). Viceversa, per $T\le 0,01$ nessuna nuova configurazione sarebbe più accettata, a meno che non accorci il tragitto del commesso. A quel punto si starebbe effettuando, di fatto, un random search tra tutti i possibili cammini. Di conseguenza, se al termine della simulazione non si sono ancora raggiunti dei buoni risultati, anziché spingersi a temperature prossime allo zero, ha più senso aumentare il numero di passi effettuati per ogni temperatura.

Per ciascuna $T$ si sono effettuati 100 step per le città sulla circonferenza e 200 per quelle nel quadrato. Per questa disposizione, infatti, si è notato che cento passi non erano sufficienti per produrre risultati soddisfacenti. In uno step, come per l'algoritmo di Metropolis, si propone una nuova configurazione e si determina se questa può essere accettata o se si deve rimanere nella precedente. 

Partendo da $T=100$, dopo i cento passi previsti, la temperatura è stata abbassata, ogni volta, di un grado fino ad arrivare a $T=5$. A quel punto si è deciso di procedere più lentamente diminuendo la temperatura di $0,1$ gradi fino alla fine del range. Così facendo, si è osservato che il decremento era sufficientemente graduale e si producevano ottimi risultati. Come nell'esercitazione precedente, il percorso iniziale è stato scelto seguendo l'ordine in cui si sono generate le città. Quando la temperatura veniva variata si ripartiva, ovviamente, dal tragitto ottenuto al termine dei cento step della $T$ precedente.

Per proporre la nuova configurazione si sono sfruttate le mutazioni già implementate nell'algoritmo genetico. Si sono però moltiplicate per dieci le quattro $p_{mut}$. In questo modo la permutazione di una coppia di città ha una $p=1$; lo shift ha $p=0,5$; la permutazione tra n città contigue ha $p=0,3$; l'inversione ha $p=0,3$. Si è dunque certi di proporre una nuova configurazione diversa dalla precedente. È importante osservare che questa scelta si è rivelata cruciale per la risoluzione dell'esercizio: con mosse differenti si doveva moltiplicare per cento (a volte per mille) il numero di step per ottenere dei risultati ragionevoli.

Nei grafici successivi verrà mostrata, in funzione del numero di step, la lunghezza del viaggio del commesso (sempre calcolata con $L^{(2)}$) e, al termine della simulazione, il percorso finale ottenuto.

```python
import numpy as np
import matplotlib.pyplot as plt


print("CITTA' SUL CERCHIO")

temp,L = np.loadtxt("esercizio10.1/best_cerchio.dat", usecols=(0,1), delimiter=' ', unpack='true')
gen=np.arange(len(L))
plt.title("Andamento $L^2$")
plt.xlabel("numero step")
plt.ylabel("$L^2$")
plt.plot(gen,L)
plt.grid(True)
plt.show()


x,y = np.loadtxt("esercizio10.1/percorso_cerchio.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y,marker='.',markerfacecolor="r")
plt.title("Percorso")
plt.grid(True)
plt.show()
```

```python

print("CITTA' NEL QUADRATO")

temp,L = np.loadtxt("esercizio10.1/best_quad.dat", usecols=(0,1), delimiter=' ', unpack='true')
gen=np.arange(len(L))
plt.title("Andamento $L^2$")
plt.xlabel("numero step")
plt.ylabel("$L^2$")
plt.plot(gen,L)
plt.grid(True)
plt.show()


x,y = np.loadtxt("esercizio10.1/percorso_quad.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y,marker='.',markerfacecolor="r")
plt.title("Percorso")
plt.grid(True)
plt.show()
```

Come si può notare dai disegni del percorso, si riesce a giungere a degli ottimi risultati. Per le città sul cerchio il percorso è sicuramente il migliore possibile: il commesso non ritorna mai sui suoi passi. Per quelle all'interno del quadrato si potrebbe, probabilmente, trovare un cammino leggermente migliore (analizzare l'itinerario tra le tre città in basso al centro). In ogni caso la soluzione ottenuta è senz'altro soddisfacente.

Come era lecito aspettarsi, la $L^{(2)}$ è oscillante a temperature piuttosto elevate, mentre tende a diventare monotona decrescente man meno che si riduce la $T$. Infatti, a bassa $T$, è molto poco probabile che una configurazione venga accettata, a meno che non riduca la lunghezza del viaggio.

I risultati sono, probabilmente, leggermente migliori di quelli ricavati tramite l'algoritmo genetico: con il simulated annealing si è riusciti ad arrivare al minimo assoluto del percorso per le città disposte sulla circonferenza. Inoltre, il tempo computazionale di questo esercizio è notevolmente inferiore del precedente.

## <span style="color:blue">ESERCIZIO 10.2</span>

Per sfruttare le potenzialità del calcolo parallelo, si utilizza nuovamente l'algoritmo genetico al fine di risolvere il problema del commesso viaggiatore. Questa volta sono stati introdotti quattro continenti. In ciascuno vi sono $100$ individui e la disposizione delle città è la stessa (si è scelto di metterle dentro il solito quadrato di lato $l=20$). Inoltre, i vari criteri di selezione per la riproduzione e le probabilità di mutazione sono analoghe a quelli già spiegati. Ogni $n_{migr}$ generazioni, tuttavia, i continenti si scambiano casualmente il loro migliore individuo. Lo scopo dell'esercizio è di verificare se l'introduzione di queste migrazioni migliori i risultati.

Per ciascun continente si è utilizzato un seme diverso del generatore di numeri casuali per non ottenere lo stesso percorso evolutivo.

A differenza della scorsa esercitazione, si è deciso di limitare il numero di generazioni a $n_{gen}=2000$. Con i vecchi parametri, infatti, al termine del programma tutti e quattro i continenti producevano risultati molto simili (e prossimi al minimo assoluto). Di conseguenza sarebbe stato molto difficile osservare eventuali variazioni determinate dalle migrazioni.

La migrazione è stata così implementata: si è estratto un numero intero casuale $k$ tra uno e cinque che determina il numero degli spostamenti (a coppie) tra gli individui di due continenti. A questo punto si sorteggiano due continenti differenti e si scambiano loro i commessi migliori (l'operazione è ripetuta ovviamente $k$ volte) tramite le funzioni "MPI_Send" e "MPI_Recv". Concretamente, per poter effettuare lo scambio, si sono salvati su degli array gli itinerari più brevi. Sarebbe stato molto più difficile, infatti, utilizzare le librerie MPI direttamente sulla classe "popolazione".

Per poter osservare gli effetti della migrazione, il programma è stato lanciato due volte, la prima ponendo $n_{migr}=10$, la seconda bloccando totalmente le migrazioni. Nei grafici seguenti verrà mostrato, per ciascuno dei quattro continenti, l'andamento della $L^{(2)}$ del miglior individuo in funzione delle generazioni. Al termine delle $n_{gen}$ previste viene disegnato il tragitto più breve. 

```python
print("Continente 1")
print()

gen,L = np.loadtxt("esercizio10.2/best_0.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.title("Andamento L^2")
plt.xlabel("generazione")
plt.ylabel("$L^2$")
plt.plot(gen,L,label="CON migrazioni")

gen,L = np.loadtxt("esercizio10.2/best_0_nomig.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.title("Andamento L^2")
plt.xlabel("generazione")
plt.ylabel("$L^2$")
plt.plot(gen,L,label="SENZA migrazioni")
plt.grid(True)
legend = plt.legend(loc='best',shadow=True)
plt.show()

x,y = np.loadtxt("esercizio10.2/percorso_0.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y,marker='.',markerfacecolor="r")
plt.title("Percorso CON migrazioni")
plt.grid(True)
plt.show()

x,y = np.loadtxt("esercizio10.2/percorso_0_nomig.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y,marker='.',markerfacecolor="r")
plt.title("Percorso SENZA migrazioni")
plt.grid(True)
plt.show()
```
