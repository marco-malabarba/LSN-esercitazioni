### **MODELLO DI ISING UNIDIMENSIONALE**

Disegnare, in funzione della temperatura, l'energia interna del sistema $U$, il calore specifico $C$, la magnetizzazione $M$ e la suscettività magnetica $\chi$ di singola particella


Il modello di Ising descrive l'interazione tra gli spin atomici di un materiale. Si assume che ciascun momento magnetico possa essere diretto solamente in due direzioni: in alto ($s=1$), oppure in basso ($s=-1$). Per semplificare i calcoli (e ridurre il tempo computazionale) si assume che ciascuno spin possa interagire solo con quelli più vicini. L'Hamiltoniana del modello di Ising unidimensionale risulta allora: $$ H = -J\sum_{i=1}^N s_i s_{i+1}-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) $$ 

Nell'esperimento si utilizzeranno le unità di misura naturali $k_B=1$ e $\mu_B=1$. Anche il valore di $J$ è stato fissato pari ad uno. Si noti che, poiché $J>0$, si sta simulando il comportamento ferromagnetico dei materiali (poiché i momenti magnetici tendono a rimanere allineati per minimizzare l'energia). Si pone $N=50$; questo valore è estremamente più piccolo del numero di atomi in una qualunque porzione macroscopica di materiale. Si utilizzano allora le *periodic boundary conditions*: in questo caso $s_{N+1} = s_{1}$.

Tutti gli spin vengono inizialmente determinati in modo casuale: si estraggono cinquanta numeri $x_{rand}$ uniformemente distribuiti tra zero ed uno. Se $x_{rand}<0,5$ si pone arbitrariamente $s=1$, altrimenti $s=-1$. A questo punto, per studiare il sistema, si può utilizzare l'algoritmo di Metropolis, oppure quello di Gibbs.

Gli step fondamentali dell'algoritmo di Metropolis sono analoghi a quelli descritti nella precedente esercitazione. È però differente la distribuzione di probabilità. Nel sistema microcanonico, la probabilità che un sistema $S$ (all'equilibrio termodinamico con il resto dell'universo) si trovi in uno stato di energia $E_m$ è: $$P_m=\frac{e^{-\beta E_m}}{Z}$$ in cui $\beta=\frac{1}{k_BT}=\frac{1}{T}$ in unità naturali. $Z$ è invece la funzione di partizione (che è indipendente dallo stato del sistema).  

Scegliendo una distribuzione di transizione simmetrica, la probabilità di accettare la mossa proposta è $\min{\left[1;\frac{P_{\nu}}{P_{\mu}}\right]}$. $\nu; \mu$ sono, rispettivamente, stati finali e iniziali generici. Nel rapporto tra le due $P$ si elide la funzione di partizione, pertanto risulta semplicemente che $\frac{P_{\nu}}{P_{\mu}}=e^{-\beta(E_{\nu}-E_{\mu})}$. 

Data una generica configurazione, per proporre la successiva si è scelto di invertire lo spin di una delle cinquanta particelle. Si è dunque estratto un numero casuale $q$ uniformemente distribuito tra zero e uno; $i=\lceil 50q\rceil$ rappresenta l'atomo a cui si modifica il momento magnetico. Seguendo infatti la notazione introdotta nell'Hamiltoniana $i=1,2,\dots,50$. Nel calcolatore, invece, poiché i valori di $s$ sono contenuti in un array, si è dovuto sottrarre uno all'espressione precedente. Poiché la distribuzione di transizione è simmetrica, la probabilità di accettare la nuova configurazione è: $$\min{\left[1;e^{-\beta(E_{\nu}-E_{\mu})}\right]}$$

La formula precedente è piuttosto intuitiva: se l'energia dello stato finale $\nu$ risulta inferiore a quella di $\mu$ la mossa è sempre accettata; nella situazione opposta l'accettazione diminuisce al diminuire della temperatura. Per $T\to0$, infatti, il sistema si trova "congelato" nel suo ground-state.

La differenza $E_\nu-E_\mu$ si può ottenere senza doversi calcolare esplicitamente le due energie. Poiché ciascuno spin interagisce solo con i suoi primi vicini risulta che $$E_\nu-E_\mu=-Js_{i, \nu}(s_{i-1}+s_{i+1})-hs_{i, \nu}+Js_{i, \mu}(s_{i-1}+s_{i+1})+hs_{i, \mu}$$ $s_i$ è lo spin che si intende di invertire. In questo modo si risparmia parecchio tempo computazionale.

Non si ha alcun controllo sulla probabilità di accettazione, di conseguenza non si riuscirà a rispettare la regola semiempirica del $50\%$. Successivamente verrà comunque mostrata questa probabilità in funzione della temperatura e si cercherà di interpretare fisicamente i risultati ottenuti.


Per simulare il modello di Ising si può anche utilizzare l'algoritmo di Gibbs. A differenza dell'algoritmo di Metropolis, la nuova configurazione viene sempre accettata (ma non è detto che differisca da quella iniziale). Partendo da uno stato iniziale, in ciascuna mossa si pone uno spin (scelto casualmente come nel caso precedente) $s_i=\pm1$  secondo le: $$p(s'_i=\pm 1 | \left[ s_j : j\neq i\right])=\frac{1}{1+\exp{( -\beta \Delta E_{s_i=\mp1\to s_i=\pm1})}}$$

Il nuovo valore di $s_i$ è quindi indipendente da quello precedente: le probabilità dipendono unicamente dalle distribuzioni condizionate definite dalle particelle vicine alla i-esima. Scelto casualmente uno dei cinquanta atomi, è allora sufficiente estrarre un ulteriore numero $x$ uniformemente distribuito tra zero ed uno. Se $x<p(s'_i=1)$ si pone l'i-esimo spin pari ad uno, altrimenti $s=-1$. 

Come nel caso precedente, si può ridurre il tempo computazionale dell'algoritmo: nel calcolo delle $p(s'_i)$ è sufficiente valutare le differenze dell'Hamiltoniana nelle vicinanze dell'atomo selezionato. A determinare la nuova configurazione è, infatti, solo l'i-esimo momento magnetico, mentre tutti gli altri rimangono fissati.

Infine si osserva che l'algoritmo di Gibbs presenta le stesse difficoltà di quello di Metropolis (spiegate nella precedente esercitazione). Si campiona correttamente, infatti, la distribuzione di probabilità richiesta dopo una fase di equilibrazione iniziale, inoltre le configurazioni che vengono prodotte sono tra loro correlate. In seguito si discuteranno le scelte effettuate per far fronte a questi problemi.


Nell'esperimento verranno utilizzati entrambi gli algoritmi e si mostrerà che producono risultati analoghi (entro le incertezze).

Si può dimostrare che le quantità termodinamiche che si intendono studiare sono pari a: $$U(N,T) = \langle H \rangle \quad \quad C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle)$$ $$$$ $$\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle \quad \quad M(N,T)= \langle \sum_{i=1,N}s_i \rangle$$

Se si è interessati, come in questo caso, alle quantità per singola particella è sufficiente dividere per $N$ tutte le equazioni precedenti.

Si vogliono graficare queste grandezze in funzione della temperatura. Si sceglie un range $T\in[0,5;2]$. $U/N$, $C/N$ e $\chi/N$ sono studiate con un campo esterno nullo, mentre, per la $M/N$, si pone $h=0,02$.

Prima di applicare le formule è necessario verificare che il sistema si trovi all'equilibrio: la configurazione di partenza è stata scelta in modo del tutto casuale e corrisponde, in generale, a una condizione di non equilibrio. Da una temperatura di $0,5$ si è allora applicato $1000$ volte l'algoritmo di Gibbs e si è calcolata, altrettante volte, $H/N$.  

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

misura = np.loadtxt("output.en_ist.0", usecols=(0), delimiter=' ', unpack='true')
j=np.arange(len(misura))
j+=1

plt.plot(j,misura)

plt.title('Andamento energia')
plt.xlabel('Misura')
plt.ylabel('$H/N$')
plt.grid(True)
plt.show()
```

L'andamento istantaneo della $H/N$ viene salvato se la variabile "verif_eq", presente nel file di input, è pari ad uno.

Si nota che in pochissimi passi (meno di un centinaio) si raggiungere l'equilibrio. Si suppone, pertanto, che in $1000$ step il sistema riesca a equilibrarsi anche a temperature più elevate. 

Viene inserita nel programma una variabile ("restart") che, se pari ad uno, permette di ripartire dalla configurazione finale (stampata su un file) ottenuta in una precedente simulazione.

Inizialmente si è studiato il sistema alla temperatura $T=0,5$, dopodiché si è aumentata $T$, ogni volta, di $0,1$ gradi fino ad arrivare a $T=2$. Ad ogni nuova temperatura (a parte ovviamente per quella iniziale), si è partiti dalla precedente configurazione di spin, si sono applicati i $1000$ step di equilibrazione e si è poi studiato il sistema, dapprima con l'algoritmo di Gibbs e, successivamente, con quello di Metropolis.   

Nel file di input sono presenti anche le variabili "metro" e "analizza". La prima, se pari ad uno, valuta le quantità termodinamiche di interesse con l'algoritmo di Metropolis, altrimenti con quello di Gibbs. La seconda, invece, se uguale a zero, impedisce al programma di stampare sui file i risultati finali. Non sono per nulla interessanti, infatti, tutti i valori numerici che si ottengono al termine delle varie fasi di equilibrazione.

Come in tutte le esperienze precedenti, per poter calcolare gli errori sulle grandezze, si suddivide la simulazione in blocchi. Ad ogni temperatura e per entrambi gli algoritmi si è utilizzato un numero totale di passi Monte Carlo pari a $N_{tot}=10^7$ e un numero di blocchi $N_{blocchi}=100$. Questo programma è stato, probabilmente, quello che ha richiesto il maggior sforzo computazionale di tutte le esercitazioni. D'altro canto, con un numero così grande di step, ci si attendono delle incertezze piuttosto piccole. Inoltre, studiando il sistema a varie temperature, si possono comparare i risultati sperimentali con quelli attesi analiticamente.

Nei grafici sottostanti sono mostrati i risultati ottenuti con i due algoritmi e le curve teoriche attese. Come richiesto, la magnetizzazione è stata calcolata con un campo esterno $h=0,02$, mentre tutte le altre con $h=0$. 

```python
t, mis, err= np.loadtxt("ene_metro.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="red",marker='*',label="Metropolis")
t, mis, err= np.loadtxt("ene_gibbs.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="blue",marker='*',label="Gibbs")
plt.title('andamento U/N(T)')
plt.ylabel('U/N')
plt.xlabel('T')
points=500
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e,color="black",label="curva teorica")

plt.grid(True)
legend = plt.legend(loc='best',shadow=True)

plt.show()

t, mis, err= np.loadtxt("heat_metro.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="red",marker='*',label="Metropolis")
t, mis, err= np.loadtxt("heat_gibbs.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="blue", marker='*',label="Gibbs")
plt.title('andamento C(T)')
plt.ylabel('C')
plt.xlabel('T')
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat,color="black",label="curva teorica")

plt.grid(True)
legend = plt.legend(loc='best',shadow=True)

plt.show()

t, mis, err= np.loadtxt("mag_metro.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="red",marker='*',label="Metropolis")
t, mis, err= np.loadtxt("mag_gibbs.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="blue",marker='*',label="Gibbs")
plt.title('andamento M(T)')
plt.xlabel('T')
plt.ylabel('M')

h=0.02
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M,color="black",label="curva teorica")
plt.grid(True)
legend = plt.legend(loc='best',shadow=True)

plt.show()

t, mis, err= np.loadtxt("chi_metro.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="red",marker='*',label="Metropolis")
t, mis, err= np.loadtxt("chi_gibbs.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(t,mis,yerr=err,color="blue",marker='*',label="Gibbs")
plt.title('andamento $\chi$(T)')
plt.xlabel('T')
plt.ylabel('$\chi$')

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X, color="black", label="curva teorica")
plt.grid(True)
legend = plt.legend(loc='best',shadow=True)

plt.show()

```

I risultati sono estremamente buoni: i valori ottenuti con i due algoritmi sono sempre perfettamente compatibili tra loro e con la curva teorica nera. Inoltre, come ci si poteva attendere, le incertezze sono estremamente contenute (è quasi impossibile riuscire a notarle nei grafici). Ciò fornisce un'ulteriore prova sull'incredibile efficacia dell'algoritmo di Metropolis (e, in questo caso, anche di quello di Gibbs).

Sulla base di questi grafici, si può affermare con una certa sicurezza che, se si aumentasse ulteriormente la temperatura, si otterrebbero dei risultati altrettanto compatibili con le funzioni attese. Ciò non è necessariamente vero, invece, per $T$ molto vicine allo zero. Questi algoritmi, in condizioni così estreme, potrebbero fornire dei valori che si discostano da quanto previsto dal calcolo analitico.

Infine, al variare della temperatura, si è anche graficato l'andamento dell'accettazione nel Metropolis:

```python
t, acc= np.loadtxt("acc.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(t,acc)
plt.title('Andamento accettazione')
plt.xlabel('T')
plt.ylabel('accettazione')
plt.grid(True)
plt.show()
```

All'aumentare della temperatura la percentuale di mosse accettate aumenta sensibilmente: a temperatura $T=0,5$ è inferiore al $5\%$, mentre a $T=2$ supera il $50\%$. Questo andamento è piuttosto intuitivo, per $T\to0$, infatti, si ha che $\beta\to+\infty$. Il sistema risulta allora "congelato" nel suo ground-state: qualunque mossa che lo allontani dallo stato fondamentale possiede una probabilità estremamente bassa di essere accettata ($e^{-x}\to 0$ per $x\to+\infty$). Viceversa, a temperature estremamente elevate, $\beta\to0$, di conseguenza tutti gli stati risultano essere equivalenti. Una qualunque mossa ha dunque una probabilità piuttosto alta di essere accettata. In questo limite, infatti, si afferma che la funzione di partizione conta semplicemente il numero di stati quantistici accessibili al sistema.
