### **Calcolare con tecniche Monte Carlo il prezzo di opzioni europee**


In un'opzione europea un venditore (*writer*) concede ad un compratore (*holder*) il diritto di comprare (**opzione call**) o di vendere (**opzione put**) un bene, ad un prezzo prestabilito $K$, in un preciso momento futuro $T$.

L'holder esercita il suo diritto se ne ricava un qualche profitto: se il prezzo del bene al tempo $T$ è pari a $S(T)$, il guadagno, per un'opzione call, è pari a $\max[0;S(T)-K]$. Infatti, se $S(T)$ dovesse risultare inferiore a $K$, l'holder deciderebbe di comprare il prodotto ad un prezzo migliore sul mercato, anziché dal writer. Seguendo analoghi ragionamenti si conclude che, per un'opzione put, il profitto è dato da $\max[0;K-S(T)]$.

Questo tipo di contratto è chiaramente asimmetrico: l'holder ottiene un diritto, mentre il writer si assume un obbligo. Per questo motivo, il writer, prima di mettersi in questa posizione di svantaggio, richiede un compenso economico. Occorre quindi determinare, tramite simulazioni numeriche, quale sia il costo equo per le opzioni europee. 

Il prezzo, per una call, è intuitivamente pari a $C=\langle e^{-rT} \max[0; S(T)-K] \rangle$, ovvero al valor medio del profitto ottenuto dall'holder ridotto di un fattore $e^{-rT}$. Una banca, infatti, fornisce un tasso di interessi pari privo di rischio $r$. Per un'opzione put, invece, si ha che: $P=\langle e^{-rT} \max[0; K-S(T)] \rangle$.

L'evoluzione del prezzo di un bene nel tempo $S(T)$ si può determinare tramite la formula ricorsiva: $$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$ Nella formula precedente $Z_i$ è una variabile casuale distribuita secondo una gaussiana con valore medio $0$ e deviazione standard $1$. $\sigma$ è la volatilità del mercato (misura la forza delle fluttuzioni statistiche del mercato), $\mu$ è il drift (misura l'incremento medio del prezzo di un bene). Inoltre, l'equazione è priva di errori di discretizzazione.

È quindi possibile stimare il prezzo delle opzioni con semplici simulazioni Monte Carlo: fissato il prezzo iniziale $S(0)$, si può valutare un gran numero di volte $S(T)$, da cui si ricava $C$ (oppure $P$) tramite le relazioni precedenti.

I costi $C$ e $P$ sono stati stimati in due modi: inizialmente calcolando direttamente il prezzo del bene all'istante $T$; successivamente suddividendo l'intervallo temporale $[0;T]$ in $100$. In questo secondo caso, $S(t)$ segue un percorso discretizzato: da $S(0)$ si calcola $S(t_1)$, $S(t_2) \dots$ e così via, fino a $S(t_{100}=T)$.

Nelle quattro simulazioni si sono utilizzati $M=10^5$ step. Un passo Monte Carlo corrisponde, in questo caso, ad una stima di $e^{-rT} \max[0; S(T)-K]$ nel caso di una call (per un'opzione put invece $e^{-rT} \max[0; K-S(T)]$). Come in tutte le esercitazioni precedenti, per poter valutare le incertezze sui risultati ottenuti, si sono suddivisi in blocchi ($N=100$) i quattro esperimenti. In ciascun blocco è, quindi, contenuto un prezzo dell'opzione, calcolato tramite la media aritmetica delle $L=\frac{M}{N}$ stime di $C$ (o $P$). Le incertezze vengono poi determinate tramite la deviazione standard sulla media.

Nei quattro grafici successivi verrà mostrato l'andamento del prezzo dell'opzione all'aumentare dei passi Monte Carlo. Analizzando inizialmente solo il primo blocco, poi i primi due e così via fino ad includerli tutti e cento, il numero di step considerati è pari a $L\times N_c$ e compre un range $[L=1000; M=10^5]$.

In queste semplici opzioni europee, il prezzo equo può essere calcolato analiticamente mediante le formule di Black e Scholes. Di conseguenza, verranno graficati (in rosso) anche questi valori "esatti" di $C$ e di $P$, in modo da poterli confrontare con i risultati ottenuti tramite la simulazione numerica. 



Si sono utilizzati i seguenti parametri:
- $S(0)=100$ (prezzo del bene al tempo iniziale in unità arbitrarie)
- $T=1$ 
- $K=100$
- $\mu=r=0.1$
- $\sigma=0.25$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

call=14.975790778311286 #da Black-Scholes
put=5.4595325819072364  #da Black-Scholes

i, costo, err = np.loadtxt("calldiretta.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,costo,yerr=err)
plt.title("Costo opzione call")
plt.xlabel('numero step')
plt.ylabel('costo')
plt.grid(True)
N = 100
x = np.linspace(0.,100000.,N,endpoint=True)
y=np.zeros(N)
y=y+call
plt.plot(x,y,color="red") 
plt.show()

i, costo, err = np.loadtxt("putdiretta.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,costo,yerr=err)
plt.title("Costo opzione put")
plt.xlabel('numero step')
plt.ylabel('costo')
plt.grid(True)
N = 100
x = np.linspace(0.,100000.,N,endpoint=True)
y=np.zeros(N)
y=y+put
plt.plot(x,y,color="red") 
plt.show()

i, costo, err = np.loadtxt("calldiscreta.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,costo,yerr=err)
plt.title("Costo opzione call con discretizzazione intervallo temporale")
plt.xlabel('numero step')
plt.ylabel('costo')
plt.grid(True)
N = 100
x = np.linspace(0.,100000.,N,endpoint=True)
y=np.zeros(N)
y=y+call
plt.plot(x,y,color="red") 
plt.show()

i, costo, err = np.loadtxt("putdiscreta.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(i,costo,yerr=err)
plt.title("Costo opzione put con discretizzazione intervallo temporale")
plt.xlabel('numero step')
plt.ylabel('costo')
plt.grid(True)
N = 100
x = np.linspace(0.,100000.,N,endpoint=True)
y=np.zeros(N)
y=y+put
plt.plot(x,y,color="red") 
plt.show()
```

I risultati sono soddisfacenti: i quattro costi ottenuti considerando tutti i 100 blocchi sono compatibili, entro le loro incertezze, con i valori previsti dalle formule di Black e Scholes. Inoltre, anche con un numero limitato di step, si osserva qualitativamente che tutti i dati non si discostano mai di oltre $2\sigma$ dal valore teorico.

È interessante notare che le incertezze sui costi, ricavate con un calcolo diretto di $S(T)$, sono estremamente simili a quelle ottenute discretizzando l'intervallo temporale $[0;T]$. Per le call si ha $\sigma\simeq0,06$, mentre per le put $\sigma\simeq0,03$. Ciò potrebbe sorprendere: sarebbe logico attendersi che, dividendo $[0;T]$ in un numero maggiori di parti, si stimi più precisamente $S(T)$ e, di conseguenza, il costo dell'opzione. Tuttavia, come già accennato, nella formula precedentemente esposta per il calcolo di $S(t_{i+1})$, non vengono introdotti errori di discretizzazione. I due metodi utilizzati sono quindi equivalenti. La discretizzazione diventa invece cruciale se si vuole calcolare il prezzo di opzioni più esotiche (ad esempio le americane e le Bermuda), in cui l'holder può decidere di esercitare il suo diritto prima di $T$.
